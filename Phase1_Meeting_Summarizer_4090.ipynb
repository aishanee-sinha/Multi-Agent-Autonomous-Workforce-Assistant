````python
# Cell 3: Flexible AMI Parser with Auto-Detection
# Description: Enhanced parser that auto-detects AMI directory structure

class FlexibleAMIParser:
    """Enhanced AMI parser that automatically detects directory structure"""

    def __init__(self, base_dir: str):
        self.base_dir = Path(base_dir)
        self.words_dir = None
        self.abstractive_dir = None
        self.meeting_ids = []
        self._detect_structure()

    def _detect_structure(self):
        """Auto-detect where the AMI files are located"""
        print(f"Auto-detecting AMI structure in: {self.base_dir}")

        # Search for .words.xml files recursively
        words_files = list(self.base_dir.glob('**/*.words.xml'))

        if not words_files:
            print("ERROR: No .words.xml files found")
            return

        # Determine words directory from first file
        first_file = words_files[0]
        self.words_dir = first_file.parent

        print(f"Detected words directory: {self.words_dir}")

        # Look for abstractive summaries
        parent = self.words_dir.parent

        possible_abstractive = [
            parent / 'abstractive',
            parent / 'abstracts',
            parent / 'summaries',
            self.words_dir,
        ]

        for abs_dir in possible_abstractive:
            if abs_dir.exists():
                abssumm_files = list(abs_dir.glob('*.abssumm.xml'))
                if abssumm_files:
                    self.abstractive_dir = abs_dir
                    print(f"Detected abstractive directory: {self.abstractive_dir}")
                    print(f"  Found {len(abssumm_files)} summary files")
                    break

        if not self.abstractive_dir:
            print("WARNING: No abstractive summaries found")
            print("  Will proceed with transcripts only")

    def find_meetings(self) -> List[str]:
        """Find all available meeting IDs"""
        if not self.words_dir:
            return []

        meeting_ids = set()

        for xml_file in self.words_dir.glob('*.words.xml'):
            parts = xml_file.stem.split('.')
            if len(parts) >= 2:
                meeting_id = parts[0]
                meeting_ids.add(meeting_id)

        self.meeting_ids = sorted(list(meeting_ids))
        print(f"Found {len(self.meeting_ids)} meetings")

        if self.meeting_ids:
            print(f"Sample meeting IDs: {self.meeting_ids[:5]}")

        return self.meeting_ids

    def parse_words(self, meeting_id: str) -> List[Dict]:
        """Parse word-level transcripts with timestamps"""
        if not self.words_dir:
            return []

        utterances = []
        speaker_files = sorted(self.words_dir.glob(f"{meeting_id}.*.words.xml"))

        if not speaker_files:
            return []

        for speaker_file in speaker_files:
            parts = speaker_file.stem.split('.')
            speaker = parts[1] if len(parts) >= 2 else 'Unknown'

            try:
                tree = etree.parse(str(speaker_file))
                root = tree.getroot()

                words = root.findall('.//{http://nite.sourceforge.net/}w')
                if not words:
                    words = root.findall('.//w')

                for word_elem in words:
                    word_text = word_elem.text or ''
                    word_text = word_text.strip()

                    if not word_text:
                        continue

                    start_time = float(word_elem.get('starttime', 0))
                    end_time = float(word_elem.get('endtime', start_time))

                    utterances.append({
                        'speaker': speaker,
                        'word': word_text,
                        'start_time': start_time,
                        'end_time': end_time
                    })

            except Exception as e:
                print(f"Error parsing {speaker_file}: {e}")
                continue

        utterances.sort(key=lambda x: x['start_time'])
        return utterances

    def group_words_to_utterances(self, words: List[Dict],
                                   pause_threshold: float = 1.0) -> List[Dict]:
        """Group words into utterances based on speaker and pauses"""
        if not words:
            return []

        utterances = []
        current_utt = {
            'speaker': words[0]['speaker'],
            'words': [words[0]['word']],
            'start_time': words[0]['start_time'],
            'end_time': words[0]['end_time']
        }

        for word in words[1:]:
            time_gap = word['start_time'] - current_utt['end_time']
            speaker_change = word['speaker'] != current_utt['speaker']

            if speaker_change or time_gap > pause_threshold:
                current_utt['text'] = ' '.join(current_utt['words'])
                utterances.append(current_utt)

                current_utt = {
                    'speaker': word['speaker'],
                    'words': [word['word']],
                    'start_time': word['start_time'],
                    'end_time': word['end_time']
                }
            else:
                current_utt['words'].append(word['word'])
                current_utt['end_time'] = word['end_time']

        if current_utt['words']:
            current_utt['text'] = ' '.join(current_utt['words'])
            utterances.append(current_utt)

        return utterances

    def parse_abstractive_summaries(self, meeting_id: str) -> Dict[str, str]:
        """Parse abstractive summaries if available"""
        summaries = {
            'abstract': '',
            'decisions': '',
            'actions': '',
            'problems': ''
        }

        if not self.abstractive_dir:
            return summaries

        summary_file = self.abstractive_dir / f"{meeting_id}.abssumm.xml"

        if not summary_file.exists():
            return summaries

        try:
            tree = etree.parse(str(summary_file))
            root = tree.getroot()

            for elem in root.findall('.//abstract'):
                text = ''.join(elem.itertext()).strip()
                summaries['abstract'] = text

            for elem in root.findall('.//decision'):
                text = ''.join(elem.itertext()).strip()
                if summaries['decisions']:
                    summaries['decisions'] += '\n' + text
                else:
                    summaries['decisions'] = text

            for elem in root.findall('.//action'):
                text = ''.join(elem.itertext()).strip()
                if summaries['actions']:
                    summaries['actions'] += '\n' + text
                else:
                    summaries['actions'] = text

            for elem in root.findall('.//problem'):
                text = ''.join(elem.itertext()).strip()
                if summaries['problems']:
                    summaries['problems'] += '\n' + text
                else:
                    summaries['problems'] = text

        except Exception as e:
            print(f"Error parsing summary for {meeting_id}: {e}")

        return summaries

    def extract_action_items(self, actions_text: str) -> List[Dict]:
        """Extract structured action items from actions text"""
        if not actions_text:
            return []

        action_items = []
        lines = actions_text.strip().split('\n')

        for line in lines:
            line = line.strip()
            if not line or len(line) < 10:
                continue

            action = {
                'task': line,
                'owner': 'TBD',
                'deadline': 'TBD',
                'confidence': 0.8
            }

            owner_match = re.search(r'\b([A-Z][a-z]+)\s+(will|should|to|needs to|has to)', line)
            if owner_match:
                action['owner'] = owner_match.group(1)

            deadline_patterns = [
                r'by\s+([^,\.]+)',
                r'before\s+([^,\.]+)',
                r'until\s+([^,\.]+)',
                r'next\s+(week|meeting|month)'
            ]

            for pattern in deadline_patterns:
                deadline_match = re.search(pattern, line, re.IGNORECASE)
                if deadline_match:
                    action['deadline'] = deadline_match.group(0)
                    break

            action_items.append(action)

        return action_items

    def parse_meeting(self, meeting_id: str) -> Dict:
        """Parse complete meeting data"""
        words = self.parse_words(meeting_id)
        utterances = self.group_words_to_utterances(words)
        summaries = self.parse_abstractive_summaries(meeting_id)
        action_items = self.extract_action_items(summaries['actions'])

        full_transcript = '\n'.join([
            f"[{utt['speaker']}] {utt['text']}"
            for utt in utterances
        ])

        return {
            'meeting_id': meeting_id,
            'utterances': utterances,
            'transcript': full_transcript,
            'summary_abstract': summaries['abstract'],
            'summary_decisions': summaries['decisions'],
            'summary_actions': summaries['actions'],
            'summary_problems': summaries['problems'],
            'action_items': action_items,
            'duration_seconds': utterances[-1]['end_time'] if utterances else 0,
            'num_utterances': len(utterances),
            'num_words': len(words)
        }

print("FlexibleAMIParser class defined")
print("This parser auto-detects AMI directory structure")
````

````python
# Cell 4: Dataset Builder Functions
# Description: Build complete dataset from AMI annotations

def build_ami_dataset_flexible(base_dir: str, output_dir: str) -> List[Dict]:
    """Build AMI dataset using flexible auto-detection"""
    parser = FlexibleAMIParser(base_dir)

    meeting_ids = parser.find_meetings()

    if not meeting_ids:
        print("\nERROR: No meetings found")
        return []

    dataset = []
    failed = []

    print(f"\nParsing {len(meeting_ids)} meetings...")

    for meeting_id in tqdm(meeting_ids, desc="Parsing"):
        try:
            meeting_data = parser.parse_meeting(meeting_id)

            if meeting_data['transcript']:
                dataset.append(meeting_data)
            else:
                failed.append(meeting_id)

        except Exception as e:
            print(f"\nFailed to parse {meeting_id}: {e}")
            failed.append(meeting_id)

    print(f"\nDataset built:")
    print(f"  Total meetings found: {len(meeting_ids)}")
    print(f"  Successfully parsed: {len(dataset)}")
    print(f"  Failed: {len(failed)}")

    with_abstract = sum(1 for m in dataset if m['summary_abstract'])
    with_actions = sum(1 for m in dataset if m['summary_actions'])

    print(f"\nSummary coverage:")
    print(f"  Meetings with abstract: {with_abstract}/{len(dataset)}")
    print(f"  Meetings with actions: {with_actions}/{len(dataset)}")

    return dataset


def create_train_val_test_split(dataset: List[Dict],
                                 train_ratio: float = 0.7,
                                 val_ratio: float = 0.15,
                                 test_ratio: float = 0.15,
                                 random_seed: int = 42) -> Dict[str, List[Dict]]:
    """Split dataset into train/val/test"""
    assert abs(train_ratio + val_ratio + test_ratio - 1.0) < 1e-6

    random.seed(random_seed)
    dataset_shuffled = dataset.copy()
    random.shuffle(dataset_shuffled)

    n = len(dataset_shuffled)
    n_train = int(n * train_ratio)
    n_val = int(n * val_ratio)

    splits = {
        'train': dataset_shuffled[:n_train],
        'val': dataset_shuffled[n_train:n_train+n_val],
        'test': dataset_shuffled[n_train+n_val:]
    }

    print(f"\nDataset splits:")
    print(f"  Train: {len(splits['train'])} meetings")
    print(f"  Val: {len(splits['val'])} meetings")
    print(f"  Test: {len(splits['test'])} meetings")

    return splits


def save_dataset_jsonl(splits: Dict[str, List[Dict]], output_dir: str):
    """Save dataset splits to JSONL files"""
    output_path = Path(output_dir)
    output_path.mkdir(parents=True, exist_ok=True)

    for split_name, split_data in splits.items():
        output_file = output_path / f"{split_name}.jsonl"

        with open(output_file, 'w', encoding='utf-8') as f:
            for meeting in split_data:
                record = {
                    'meeting_id': meeting['meeting_id'],
                    'transcript': meeting['transcript'],
                    'summary_abstract': meeting['summary_abstract'],
                    'summary_decisions': meeting['summary_decisions'],
                    'summary_actions': meeting['summary_actions'],
                    'summary_problems': meeting['summary_problems'],
                    'action_items': meeting['action_items'],
                    'duration_seconds': meeting['duration_seconds'],
                    'num_utterances': meeting['num_utterances']
                }
                f.write(json.dumps(record, ensure_ascii=False) + '\n')

        print(f"Saved {split_name}: {output_file}")

    metadata = {
        'total_meetings': sum(len(v) for v in splits.values()),
        'train_size': len(splits['train']),
        'val_size': len(splits['val']),
        'test_size': len(splits['test']),
        'created_at': datetime.now().isoformat(),
        'random_seed': cfg.RANDOM_SEED
    }

    metadata_file = output_path / 'metadata.json'
    with open(metadata_file, 'w') as f:
        json.dump(metadata, f, indent=2)

    print(f"Saved metadata: {metadata_file}")

print("Dataset builder functions defined")
````

````python
# Cell 6: Prompt Template and Training Data Preparation
# Description: Create instruction prompts for Qwen model training

def create_meeting_summary_prompt(transcript: str, include_instructions: bool = True) -> str:
    """Create instruction prompt for meeting summarization"""

    instruction = """You are an expert meeting summarizer. Given a meeting transcript, generate a comprehensive meeting summary with the following sections:

1. ABSTRACT: A brief overview of the meeting
2. DECISIONS: Key decisions made during the meeting
3. ACTIONS: Action items with owners and deadlines
4. PROBLEMS: Issues or challenges discussed

Format action items as:
[Owner] - Task description - Due: Deadline

Be concise, factual, and preserve important details."""

    if include_instructions:
        prompt = f"""<|im_start|>system
You are a helpful assistant specialized in meeting summarization.<|im_end|>
<|im_start|>user
{instruction}

TRANSCRIPT:
{transcript}

Generate the meeting summary:<|im_end|>
<|im_start|>assistant
"""
    else:
        prompt = f"""TRANSCRIPT:
{transcript}

Generate the meeting summary:
"""

    return prompt


def create_target_summary(meeting_data: Dict) -> str:
    """Create target summary from meeting data"""
    sections = []

    if meeting_data.get('summary_abstract'):
        sections.append(f"ABSTRACT:\n{meeting_data['summary_abstract']}")

    if meeting_data.get('summary_decisions'):
        sections.append(f"\nDECISIONS:\n{meeting_data['summary_decisions']}")

    if meeting_data.get('action_items') and len(meeting_data['action_items']) > 0:
        actions_text = "\nACTIONS:"
        for action in meeting_data['action_items']:
            owner = action.get('owner', 'TBD')
            task = action.get('task', '')
            deadline = action.get('deadline', 'TBD')
            actions_text += f"\n[{owner}] - {task} - Due: {deadline}"
        sections.append(actions_text)
    elif meeting_data.get('summary_actions'):
        sections.append(f"\nACTIONS:\n{meeting_data['summary_actions']}")

    if meeting_data.get('summary_problems'):
        sections.append(f"\nPROBLEMS:\n{meeting_data['summary_problems']}")

    summary = '\n'.join(sections)
    summary += "<|im_end|>"

    return summary


def prepare_training_dataset(jsonl_file: str, tokenizer, max_length: int = 4096) -> Dataset:
    """Prepare dataset for training from JSONL file"""
    data = []
    with open(jsonl_file, 'r', encoding='utf-8') as f:
        for line in f:
            data.append(json.loads(line))

    print(f"Loaded {len(data)} samples from {jsonl_file}")

    samples = []
    skipped = 0

    for meeting in tqdm(data, desc="Preparing samples"):
        prompt = create_meeting_summary_prompt(meeting['transcript'], include_instructions=True)
        target = create_target_summary(meeting)
        full_text = prompt + target

        tokenized = tokenizer(
            full_text,
            truncation=True,
            max_length=max_length,
            padding=False,
            return_tensors=None
        )

        if len(tokenized['input_ids']) >= max_length:
            skipped += 1
            continue

        # Only store tensor-compatible fields
        samples.append({
            'input_ids': tokenized['input_ids'],
            'attention_mask': tokenized['attention_mask']
        })

    print(f"Prepared {len(samples)} samples ({skipped} skipped due to length)")

    dataset = Dataset.from_list(samples)
    return dataset


def show_training_sample(dataset: Dataset, tokenizer, idx: int = 0):
    """Display a training sample for verification"""
    sample = dataset[idx]

    print("=" * 80)
    print(f"TRAINING SAMPLE {idx}")
    print("=" * 80)
    print(f"Input IDs length: {len(sample['input_ids'])}")

    decoded_text = tokenizer.decode(sample['input_ids'], skip_special_tokens=False)

    print(f"\nFull Text (first 1000 chars):")
    print(decoded_text[:1000])
    print("\n...")
    print(decoded_text[-500:])
    print("=" * 80)

print("Prompt template and data preparation functions defined")
````

##Evaluation

````python
# Cell 10: Evaluation Functions (Fixed)
# Description: G-Eval metrics for comprehensive evaluation

# Download required NLTK data
import nltk
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)  # Additional requirement for newer NLTK

def _normalize(x, lo=0.0, hi=1.0):
    """Normalize value to range"""
    return max(lo, min(hi, x))

def _ratio(n, d):
    """Safe division ratio"""
    return 0.0 if d <= 0 else n / d

def _tokenize(s: str) -> List[str]:
    """Simple word tokenization"""
    return re.findall(r"\w+", s.lower())

def geval_coherence(summary: str) -> float:
    """Measure logical flow and coherence"""
    sents = [s for s in sent_tokenize(summary) if s.strip()]
    if not sents:
        return 0.0

    avg_len = sum(len(_tokenize(s)) for s in sents) / len(sents)
    penalty_frag = 1.0 if avg_len >= 8 else avg_len / 8.0

    overlaps = []
    for i in range(len(sents) - 1):
        a = set(_tokenize(sents[i]))
        b = set(_tokenize(sents[i + 1]))
        overlaps.append(_ratio(len(a & b), len(a | b)))

    local = np.mean(overlaps) if overlaps else 0.0
    return _normalize(0.6 * penalty_frag + 0.4 * local)

def geval_consistency(summary: str, source: str) -> float:
    """Measure factual consistency with source"""
    src = set(_tokenize(source))
    summ = _tokenize(summary)

    numbers = [t for t in summ if re.fullmatch(r"\d+", t)]
    proper_like = [p.lower() for p in re.findall(r"\b[A-Z][a-zA-Z]+\b", summary)]

    hallu = sum(1 for t in numbers + proper_like if t not in src)
    rate = 1.0 - _ratio(hallu, max(1, len(numbers) + len(proper_like)))
    return _normalize(rate)

def geval_fluency(summary: str) -> float:
    """Measure language quality and fluency"""
    toks = _tokenize(summary)
    sents = [s for s in sent_tokenize(summary) if s.strip()]
    if not sents:
        return 0.0

    avg_len = len(toks) / len(sents)
    rep_pen = 1.0 - _ratio(len(toks) - len(set(toks)), max(1, len(toks)))
    len_score = 1.0 - min(1.0, abs(avg_len - 20) / 20)

    return _normalize(0.5 * rep_pen + 0.5 * len_score)

def geval_relevance(summary: str, source: str) -> float:
    """Measure relevance to source content"""
    src_toks = set(_tokenize(source))
    summ_toks = _tokenize(summary)

    if not summ_toks:
        return 0.0

    relevant_words = sum(1 for t in summ_toks if t in src_toks)
    coverage = _ratio(relevant_words, len(summ_toks))

    src_freq = {}
    for t in _tokenize(source):
        src_freq[t] = src_freq.get(t, 0) + 1

    top_src = sorted(src_freq.items(), key=lambda x: x[1], reverse=True)[:20]
    top_words = set([w for w, _ in top_src])

    top_coverage = sum(1 for t in summ_toks if t in top_words)
    top_score = _ratio(top_coverage, len(summ_toks))

    return _normalize(0.6 * coverage + 0.4 * top_score)

def geval_action_quality(summary: str) -> float:
    """Measure quality of action items extraction"""
    action_patterns = [
        r"(?i)action\s*items?:",
        r"(?i)to-?do:",
        r"(?i)follow-?ups?:",
        r"(?i)next\s+steps?:"
    ]

    has_section = any(re.search(pat, summary) for pat in action_patterns)

    if not has_section:
        return 0.0

    action_markers = len(re.findall(r"^\s*[-•*]\s+", summary, re.MULTILINE))
    action_verbs = len(re.findall(
        r"\b(will|should|must|need|needs|plan|plans|planning|schedule|scheduled)\b",
        summary,
        re.IGNORECASE
    ))

    structure_score = min(1.0, action_markers / 3.0)
    verb_score = min(1.0, action_verbs / 5.0)

    return _normalize(0.6 * structure_score + 0.4 * verb_score)

def geval_comprehensive(summary: str, transcript: str) -> Dict[str, float]:
    """Calculate all evaluation metrics"""
    coherence = geval_coherence(summary)
    consistency = geval_consistency(summary, transcript)
    fluency = geval_fluency(summary)
    relevance = geval_relevance(summary, transcript)
    action_quality = geval_action_quality(summary)

    geval_overall = np.mean([coherence, consistency, fluency, relevance, action_quality])

    production_score = (
        0.25 * coherence +
        0.25 * consistency +
        0.20 * fluency +
        0.20 * relevance +
        0.10 * action_quality
    )

    return {
        'coherence': coherence,
        'consistency': consistency,
        'fluency': fluency,
        'relevance': relevance,
        'action_quality': action_quality,
        'geval_overall': geval_overall,
        'production_score': production_score
    }

print("Evaluation functions defined")
print("NLTK punkt_tab downloaded")
print("Metrics: Coherence, Consistency, Fluency, Relevance, Action Quality")
````

````python
# Cell 11: Inference Function
# Description: Generate summaries using the fine-tuned model

def generate_summary(model, tokenizer, transcript: str, max_length: int = 1024) -> str:
    """Generate summary from transcript using fine-tuned model"""

    prompt = create_meeting_summary_prompt(transcript, include_instructions=True)

    inputs = tokenizer(
        prompt,
        max_length=cfg.MAX_SEQ_LENGTH,
        truncation=True,
        return_tensors="pt",
    ).to(model.device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_length,
            min_new_tokens=50,
            num_beams=4,
            length_penalty=1.0,
            no_repeat_ngram_size=3,
            early_stopping=True,
            temperature=0.7,
            do_sample=False
        )

    full_output = tokenizer.decode(outputs[0], skip_special_tokens=False)

    if "<|im_start|>assistant" in full_output:
        summary = full_output.split("<|im_start|>assistant")[-1]
        summary = summary.split("<|im_end|>")[0]
    else:
        summary = tokenizer.decode(outputs[0], skip_special_tokens=True)
        if prompt in summary:
            summary = summary[len(prompt):].strip()

    return summary.strip()

print("Inference function defined")
print("Ready to generate summaries on test set")
````

````python
# Cell 12: Run Evaluation on Test Set
# Description: Generate summaries and calculate metrics for all test samples

import time

# Load test data
test_data = []
with open(os.path.join(cfg.OUTPUT_DIR, 'test.jsonl'), 'r', encoding='utf-8') as f:
    for line in f:
        test_data.append(json.loads(line))

print(f"Loaded {len(test_data)} test samples")
print(f"Starting evaluation...")
print("=" * 80)

# Run evaluation
results = []
eval_start = time.time()

for idx, item in enumerate(tqdm(test_data, desc="Evaluating")):
    transcript = item['transcript']
    meeting_id = item['meeting_id']

    t0 = time.time()
    summary = generate_summary(model, tokenizer, transcript, max_length=1024)
    gen_time = time.time() - t0

    metrics = geval_comprehensive(summary, transcript)
    metrics['time_sec'] = gen_time
    metrics['summary'] = summary
    metrics['meeting_id'] = meeting_id
    metrics['transcript'] = transcript

    results.append(metrics)

    if (idx + 1) % 10 == 0:
        print(f"\nProcessed {idx + 1}/{len(test_data)} samples")
        print(f"  Last sample time: {gen_time:.2f}s")
        print(f"  Average production score so far: {np.mean([r['production_score'] for r in results]):.3f}")

eval_time = time.time() - eval_start

print("\n" + "=" * 80)
print("EVALUATION COMPLETE")
print("=" * 80)
print(f"Total time: {eval_time:.2f}s")
print(f"Average time per sample: {eval_time / len(test_data):.2f}s")
print(f"Samples processed: {len(results)}")
````

````python
# Cell 13: Aggregate and Display Results
# Description: Calculate aggregate statistics and display results

def aggregate_metrics(results_list):
    """Aggregate metrics across all samples"""
    metrics_keys = [
        "coherence", "consistency", "fluency", "relevance",
        "action_quality", "geval_overall", "production_score", "time_sec"
    ]

    agg = {}
    for key in metrics_keys:
        values = [r[key] for r in results_list]
        agg[f"{key}_mean"] = np.mean(values)
        agg[f"{key}_std"] = np.std(values)
        agg[f"{key}_min"] = np.min(values)
        agg[f"{key}_max"] = np.max(values)

    return agg

# Calculate aggregate statistics
agg_results = aggregate_metrics(results)

# Display results
print("=" * 80)
print("FINE-TUNED MODEL EVALUATION RESULTS")
print("=" * 80)
print(f"\nPerformance Metrics (mean +/- std):")
print(f"  Coherence:        {agg_results['coherence_mean']:.3f} +/- {agg_results['coherence_std']:.3f}")
print(f"  Consistency:      {agg_results['consistency_mean']:.3f} +/- {agg_results['consistency_std']:.3f}")
print(f"  Fluency:          {agg_results['fluency_mean']:.3f} +/- {agg_results['fluency_std']:.3f}")
print(f"  Relevance:        {agg_results['relevance_mean']:.3f} +/- {agg_results['relevance_std']:.3f}")
print(f"  Action Quality:   {agg_results['action_quality_mean']:.3f} +/- {agg_results['action_quality_std']:.3f}")
print(f"\nOverall Metrics:")
print(f"  G-Eval Overall:   {agg_results['geval_overall_mean']:.3f} +/- {agg_results['geval_overall_std']:.3f}")
print(f"  Production Score: {agg_results['production_score_mean']:.3f} +/- {agg_results['production_score_std']:.3f}")
print(f"\nInference Statistics:")
print(f"  Avg Time:         {agg_results['time_sec_mean']:.2f}s")
print(f"  Min Time:         {agg_results['time_sec_min']:.2f}s")
print(f"  Max Time:         {agg_results['time_sec_max']:.2f}s")

# Save results
results_file = os.path.join(cfg.OUTPUT_DIR, 'evaluation_results.json')
with open(results_file, 'w') as f:
    results_to_save = []
    for r in results:
        r_copy = r.copy()
        r_copy['summary'] = r_copy['summary'][:500]
        r_copy['transcript'] = r_copy['transcript'][:500]
        results_to_save.append(r_copy)

    json.dump({
        'aggregate': agg_results,
        'individual_results': results_to_save,
        'num_samples': len(results)
    }, f, indent=2)

print(f"\nResults saved to: {results_file}")
````

````python
# Cell 15: Display Sample Outputs
# Description: Show example summaries from test set

# Find samples with action items
samples_with_actions = []
for idx, result in enumerate(results):
    if 'ACTIONS:' in result['summary'] or 'Action Items:' in result['summary']:
        action_section = re.search(
            r'ACTIONS?:?\s*(.*?)(?=\n\n|PROBLEMS?:|$)',
            result['summary'],
            re.DOTALL | re.IGNORECASE
        )
        if action_section:
            action_text = action_section.group(1).strip()
            action_lines = [
                line for line in action_text.split('\n')
                if line.strip() and (
                    line.strip().startswith('[') or
                    line.strip().startswith('-') or
                    line.strip().startswith('•')
                )
            ]
            if action_lines:
                samples_with_actions.append({
                    'idx': idx,
                    'num_actions': len(action_lines),
                    'production_score': result['production_score']
                })

samples_with_actions.sort(key=lambda x: x['production_score'], reverse=True)

print("=" * 80)
print("SAMPLE OUTPUTS")
print("=" * 80)
print(f"\nFound {len(samples_with_actions)} samples with action items")
print(f"Showing top 3 samples by production score\n")

num_samples = min(3, len(samples_with_actions)) if samples_with_actions else min(3, len(results))

for i in range(num_samples):
    if samples_with_actions:
        sample_info = samples_with_actions[i]
        sample_idx = sample_info['idx']
    else:
        sorted_by_score = sorted(enumerate(results), key=lambda x: x[1]['production_score'], reverse=True)
        sample_idx = sorted_by_score[i][0]

    result = results[sample_idx]

    print("=" * 80)
    print(f"SAMPLE {i+1}")
    print("=" * 80)
    print(f"Meeting ID: {result['meeting_id']}")
    print(f"Production Score: {result['production_score']:.3f}")
    print(f"Generation Time: {result['time_sec']:.2f}s")

    print(f"\nMetrics:")
    print(f"  Coherence:      {result['coherence']:.3f}")
    print(f"  Consistency:    {result['consistency']:.3f}")
    print(f"  Fluency:        {result['fluency']:.3f}")
    print(f"  Relevance:      {result['relevance']:.3f}")
    print(f"  Action Quality: {result['action_quality']:.3f}")

    print(f"\nTranscript (first 300 chars):")
    print(result['transcript'][:300] + "...")

    print(f"\nGenerated Summary:")
    print("-" * 80)
    print(result['summary'])
    print("-" * 80)
    print()

# Show statistics by score range
print("\n" + "=" * 80)
print("PERFORMANCE BREAKDOWN BY SCORE RANGE")
print("=" * 80)

score_ranges = [
    ("Excellent", 0.8, 1.0),
    ("Good", 0.6, 0.8),
    ("Fair", 0.4, 0.6),
    ("Poor", 0.0, 0.4)
]

for label, low, high in score_ranges:
    count = sum(1 for r in results if low <= r['production_score'] < high)
    percentage = (count / len(results)) * 100
    print(f"{label:12s} ({low:.1f}-{high:.1f}): {count:3d} samples ({percentage:5.1f}%)")

print("\n" + "=" * 80)
````
