Read JIRA issues

```python
import requests
from requests.auth import HTTPBasicAuth
from google.colab import userdata
import os


email = "aishanee.sinha@sjsu.edu"
api_token = userdata.get("JIRA_API_TOKEN")
site = "https://sjsuteam5.atlassian.net"
project_key = "KAN"

url = f"{site}/rest/api/3/search/jql"
params = {
    "jql": f"project={project_key}",
    "maxResults": 50,
    "fields": "summary,description,assignee,status,issuetype"
}

response = requests.get(url,
                        auth=HTTPBasicAuth(email, api_token),
                        headers={"Accept": "application/json"},
                        params=params)

data = response.json()

for issue in data.get("issues", []):
    key = issue["key"]
    fields = issue["fields"]
    summary = fields.get("summary")
    description = fields.get("description")
    assignee = fields.get("assignee")
    assignee_name = assignee.get("displayName") if assignee else "Unassigned"
    status = fields.get("status", {}).get("name")
    issuetype = fields.get("issuetype", {}).get("name")
    comments = fields.get("comment", {}).get("comments")
    created_date = fields.get("created")
    creation = created_date.split("T")[0] if created_date else "N/A"
    resolution = fields.get("resolution", {}).get("name") if fields.get("resolution") else "N/A"
    priority = fields.get("priority", {}).get("name") if fields.get("priority") else "N/A"


    print(f"Issue: {key}")
    print(f"Summary: {summary}")
    print(f"Description: {description}")
    print(f"Assignee: {assignee_name}")
    print(f"Status: {status}")
    print(f"Issue Type: {issuetype}")
    print(f"Comments: {comments}")
    print(f"Issue Creation Date: {creation}")
    print(f"Issue Resolution Date: {resolution}")
    print(f"Priority: {priority}")
    print("-" * 40)
```

Create JIRA Issues

```python
import requests
from requests.auth import HTTPBasicAuth

email = "aishanee.sinha@sjsu.edu"
api_token=userdata.get('JIRA_API_TOKEN')
site = "https://sjsuteam5.atlassian.net"
project_key = "KAN"
account_id = "712020:7c4cf37a-d64a-4996-9058-554d93aaef95"

url = f"{site}/rest/api/3/issue"

payload = {
    "fields": {
        "project": {"key": project_key},
        "summary": "Sample Issue created using API",
        "issuetype": {"name": "Task"},
        "assignee": {"id": account_id},
        "description": {
            "type": "doc",
            "version": 1,
            "content": [
                {
                    "type": "paragraph",
                    "content": [
                        {"type": "text", "text": "Details: Sample issue created using API."}
                    ]
                }
            ]
        }
    }
}

response = requests.post(url,
                         json=payload,
                         auth=HTTPBasicAuth(email, api_token),
                         headers={"Accept": "application/json", "Content-Type": "application/json"})

print(response.status_code)
print(response.text)
```
