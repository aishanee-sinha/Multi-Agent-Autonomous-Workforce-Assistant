```python

import pandas as pd
from email import policy
from email.parser import Parser
from email.header import decode_header
from email.utils import parsedate_to_datetime, getaddresses
from bs4 import BeautifulSoup
import re

SUBJ_PREFIX = re.compile(r"^\s*((re|fwd|fw)\s*:\s*)+", re.I)
QUOTE_SPLIT_RE = re.compile(
    r"\n\s*-----\s*Original Message\s*-----\s*|\n\s*On .*? wrote:|\n\s*From:\s*.*?\s*\[mailto:.*?\]",
    re.IGNORECASE | re.DOTALL
)

# --- NEW: Regex for removing common signatures and disclaimers ---
SIGNATURE_RE = re.compile(
    r"(?s)"  # The (?s) flag makes '.' match newlines
    r"("
    # Common sign-offs (must be at the start of a new line)
    r"\n--\s*|\n__\s*|\n\s*(Best|Kind|Warm)\s*Regards,?"
    r"|\n\s*(Sincerely|Thanks|Cheers),?"
    # Common legal disclaimers
    r"|\n\s*This e-mail and any attachments.*"
    r"|\n\s*CONFIDENTIALITY(?: NOTICE| WARNING):.*"
    r")"
    # Match everything that follows until the end of the string
    r".*$"
)

def _dec(h):
    if not h: return ""
    parts = []
    for frag, enc in decode_header(h):
        parts.append(frag.decode(enc or "utf-8", "ignore") if isinstance(frag, bytes) else frag)
    return "".join(parts)

def _canon_subject(s):
    s = _dec(s or "")
    s = SUBJ_PREFIX.sub("", s).strip().lower()
    return re.sub(r"\s+", " ", s)

def _addr_list(h):
    return [e.lower() for _, e in getaddresses([_dec(h or "")]) if e]

def _html2txt(html):
    return BeautifulSoup(html or "", "lxml").get_text("\n", strip=True)

def _body(msg):
    if msg.is_multipart():
        plain, html = None, None
        for p in msg.walk():
            ct = p.get_content_type()
            if ct == "text/plain" and plain is None:
                try:
                    plain = p.get_content()
                except:
                    plain = (p.get_payload(decode=True) or b"").decode(p.get_content_charset() or "utf-8", "ignore")
            elif ct == "text/html" and html is None:
                try:
                    html = p.get_content()
                except:
                    html = (p.get_payload(decode=True) or b"").decode(p.get_content_charset() or "utf-8", "ignore")
        if plain and plain.strip():
            return plain.strip()
        if html:
            return _html2txt(html)
        return ""
    content = msg.get_content()
    return _html2txt(content) if msg.get_content_type() == "text/html" else (content or "").strip()

def _clean_body_advanced(text):
    """
    Cleans the email body by removing quoted replies, signatures, and disclaimers.
    """
    if not text:
        return ""

    text = QUOTE_SPLIT_RE.split(text, 1)[0]

    text = SIGNATURE_RE.sub("", text)
    text = re.sub(r'\n{3,}', '\n\n', text)

    return text.strip()

def parse_message_min(raw):
    try:
        msg = Parser(policy=policy.default).parsestr(raw or "")
        body_full = _body(msg)

        body_curr = _clean_body_advanced(body_full)

        dt = parsedate_to_datetime(_dec(msg.get("Date"))) if msg.get("Date") else None
        tzoff = int(dt.utcoffset().total_seconds() // 60) if (dt and dt.utcoffset()) else None

        return {
            "msg_id": _dec(msg.get("Message-ID")) or None,
            "in_reply_to": _dec(msg.get("In-Reply-To")) or None,
            "references": (_dec(msg.get("References")) or "").replace("\n", " ").split(),
            "date_utc": pd.to_datetime(dt, utc=True, errors="coerce"),
            "tz_offset_minutes": tzoff,
            "from_email": (_addr_list(msg.get("From")) or [None])[0],
            "to_emails": _addr_list(msg.get("To")),
            "cc_emails": _addr_list(msg.get("Cc")),
            "bcc_emails": _addr_list(msg.get("Bcc")),
            "subject_raw": _dec(msg.get("Subject")) or "",
            "subject_canon": _canon_subject(msg.get("Subject")),
            "body_full": body_full,
            "body_current": body_curr,
        }
    except:
        return {
            "msg_id": None, "in_reply_to": None, "references": [],
            "date_utc": pd.NaT, "tz_offset_minutes": None,
            "from_email": None, "to_emails": [], "cc_emails": [], "bcc_emails": [],
            "subject_raw": "", "subject_canon": "",
            "body_full": "", "body_current": "",
        }

```

```python
import regex as re

# ---- markers ----
MARK_FWD = re.compile(r"""(?xim)
    ^\s*(?:[-_]{5,}\s*original\s+message\s*[-_]{5,}|
           begin\s+forwarded\s+message|
           [-\s]{6,}\s*forwarded\s+by\s+.+?\s+on\s+\d{1,2}/\d{1,2}/\d{2,4}.*)$
""")
BANNER_TOP = re.compile(r"""(?xim)
\A\s*[-\s]{6,}\s*Forwarded\s+by\s+(?P<by>.+?)\s+on\s+(?P<date>\d{1,2}/\d{1,2}/\d{2,4})
(?:\s*\n\s*(?P<time>\d{1,2}:\d{2}\s*(?:AM|PM|am|pm)))?\s*[-\s]*\s*(?:\n|$)
""")

INLINE_HDR_BLOCK = re.compile(r"""(?xims)
(?:^|\n)(?:from:\s.*\n)(?:sent:\s.*\n)?(?:date:\s.*\n)?(?:to:\s.*\n)+(?:cc:\s.*\n)*(?:subject:\s.*\n)+(?:\n|$)
""")
HEADER_LINES = re.compile(r"^(from|sent|date|to|cc|bcc|subject|reply-to|attachments?):\s.*$", re.I|re.M)
ON_WROTE    = re.compile(r"(?s)(?:^|\n)on .+? wrote:\s*.*$", re.I)
GT_QUOTES   = re.compile(r"(?m)^\s*>\s?.*$")
SIG_SPLIT   = re.compile(r"\n(?:--\s*$|thanks[,\s]*$|regards[,\s]*$|best[,\s]*$)", re.I)

URL_RE   = re.compile(r"https?://\S+", re.I)
PHONE_RE = re.compile(r"""(?x)(?:\+?\d{1,3}[\s\-.]?)?(?:\(\d{3}\)|\d{3})[\s\-.]?\d{3}[\s\-.]?\d{4}""")
MOJIBAKE = re.compile(r"[?]{3,}")
DE_DASH  = re.compile(r"^[\-\_=]{5,}$", re.M)
MULTIBLANK = re.compile(r"\n{3,}")

MEET_CUES = re.compile(r"\b(meet|meeting|sync|catch\s?up|call|schedule|invite|reschedul\w*|calendar)\b", re.I)

def _strip_headers_block(s: str) -> str:
    prev = None
    t = s
    while prev != t:
        prev = t
        t = INLINE_HDR_BLOCK.sub("\n", t)
    return HEADER_LINES.sub("", t)

def _polish(s: str, max_chars=3000) -> str:
    t = s
    t = ON_WROTE.sub("", t)
    t = GT_QUOTES.sub("", t)
    t = URL_RE.sub("<URL>", t)
    t = PHONE_RE.sub("<PHONE>", t)
    t = MOJIBAKE.sub("", t)
    t = DE_DASH.sub("", t)
    parts = SIG_SPLIT.split(t, maxsplit=1)
    if len(parts) > 1:
        t = parts[0].rstrip()
    t = "\n".join(ln.strip() for ln in t.splitlines())
    t = MULTIBLANK.sub("\n\n", t).strip()
    if max_chars and len(t) > max_chars:
        t = t[:max_chars].rsplit("\n", 1)[0].strip()
    return t

def split_sections(raw: str):
    """Return (preface, [forwarded_blocks]), where each block is dict with 'headers','body','meta'."""
    if not raw:
        return "", []

    t = raw.replace("\r\n","\n").replace("\r","\n").strip()

    # eat top 'Forwarded by ...' banner and capture who/when
    meta_banner = {}
    m = BANNER_TOP.match(t)
    if m:
        meta_banner = {
            "forwarded_by": (m.group("by") or "").strip() or None,
            "forwarded_on_text": (" ".join(x for x in [(m.group("date") or "").strip(), (m.group("time") or "").strip()] if x) or None)
        }
        t = t[m.end():].lstrip()
    else:
        meta_banner = {"forwarded_by": None, "forwarded_on_text": None}

    # find all forwarded/original markers (line-based)
    lines = t.splitlines()
    idxs = [i for i,ln in enumerate(lines) if MARK_FWD.match(ln or "")]
    blocks = []
    if not idxs:
        return t, []

    # preface = everything before first marker
    preface = "\n".join(lines[:idxs[0]]).strip()

    # for each marker, carve out block to next marker
    idxs.append(len(lines))
    for a,b in zip(idxs[:-1], idxs[1:]):
        chunk = "\n".join(lines[a:b]).strip()
        # split headers vs body: headers run to first blank line
        if "\n\n" in chunk:
            hdr, body = chunk.split("\n\n", 1)
        else:
            # fallback: treat first ~10 lines that look like headers as hdr
            hdr_lines, body_lines = [], []
            for ln in chunk.splitlines():
                if HEADER_LINES.match(ln):
                    hdr_lines.append(ln)
                else:
                    body_lines.append(ln)
            hdr = "\n".join(hdr_lines)
            body = "\n".join(body_lines)
        blocks.append({
            "headers": hdr.strip(),
            "body": body.strip(),
            "meta": meta_banner.copy()
        })
        # only the first block gets the banner meta; clear for others
        meta_banner = {"forwarded_by": None, "forwarded_on_text": None}

    return preface, blocks

def choose_and_clean_training_text(body_full_or_current: str, min_chars=60, max_chars=3000):
    """
    Returns (body_train_v3, meta) where meta includes:
      segment_used: 'preface' | 'forwarded_body'
      forwarded_by, forwarded_on_text, chars_used, had_meeting_cues
    """
    preface, fwd_blocks = split_sections(body_full_or_current)
    # clean candidates lightly before measuring substance
    pref_clean = _strip_headers_block(preface)
    pref_clean = _polish(pref_clean, max_chars=None)

    # is preface substantive?
    pref_substantive = (len(re.sub(r"\s+","",pref_clean)) >= min_chars) or bool(MEET_CUES.search(pref_clean))

    if pref_substantive:
        used = _polish(_strip_headers_block(pref_clean), max_chars=max_chars)
        return used, {
            "segment_used": "preface",
            "forwarded_by": None,
            "forwarded_on_text": None,
            "chars_used": len(used),
            "had_meeting_cues": bool(MEET_CUES.search(used))
        }

    # otherwise, take first substantial forwarded body
    for blk in fwd_blocks:
        fb = _polish(_strip_headers_block(blk["body"]), max_chars=None)
        if len(re.sub(r"\s+","",fb)) >= min_chars or MEET_CUES.search(fb):
            used = _polish(fb, max_chars=max_chars)
            return used, {
                "segment_used": "forwarded_body",
                "forwarded_by": blk["meta"].get("forwarded_by"),
                "forwarded_on_text": blk["meta"].get("forwarded_on_text"),
                "chars_used": len(used),
                "had_meeting_cues": bool(MEET_CUES.search(used))
            }

    # fallback: preface even if short
    used = _polish(pref_clean, max_chars=max_chars)
    return used, {
        "segment_used": "preface",
        "forwarded_by": None,
        "forwarded_on_text": None,
        "chars_used": len(used),
        "had_meeting_cues": bool(MEET_CUES.search(used))
    }
```

```python
res = df_sample["body_current"].apply(choose_and_clean_training_text)
df_sample["body_train_v3"]   = res.apply(lambda x: x[0])
meta = res.apply(lambda x: x[1])
df_sample["segment_used"]    = meta.apply(lambda m: m["segment_used"])
df_sample["forwarded_by"]    = meta.apply(lambda m: m["forwarded_by"])
df_sample["forwarded_on_text"]= meta.apply(lambda m: m["forwarded_on_text"])

```

```python
# FAST PIPELINE: keyword shortlist -> zero-shot on candidates only (no sentence-transformers)
# Outputs:
#   artifacts/emails_meeting_POS_top1000.csv
#   artifacts/scored_full.parquet
#   artifacts/labels_merged.parquet

# !pip install -q transformers torch scikit-learn pandas pyarrow
import os, re, hashlib
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_recall_curve, average_precision_score, classification_report
from transformers import pipeline

# ---------------- CONFIG ----------------
LABELS_CSV   = "email_labels.csv"     # your manually labeled ~200 rows
SUB_COL      = "subject_raw"
BODY_COL     = "body_train_v3"
IDX_COL      = "idx"                  # index saved by your UI tool
LABEL_COL    = "has_meeting"          # boolean in your labels csv

# SPEED: lighter zero-shot model (2–3× faster than bart-large)
MODEL_NAME   = "valhalla/distilbart-mnli-12-6"
HYP_LABELS   = ["meeting intent", "no meeting intent"]
HYP_TEMPLATE = "This email is about {}."

TOPK         = 1000
BATCH        = 64                     # increase if GPU VRAM allows
TARGET_PREC  = 0.90

# Shortlist size cap: if regex hits are huge, cap to this many best candidates by heuristics
CANDIDATE_CAP = 20000

# ---------------- HELPERS ----------------
def combine_text(df: pd.DataFrame) -> pd.Series:
    return (df[SUB_COL].fillna("") + "\n\n" + df[BODY_COL].fillna("")).astype(str).str.strip()

def make_hash(s: pd.Series) -> pd.Series:
    return s.apply(lambda x: hashlib.sha1(x.encode("utf-8","ignore")).hexdigest())

def zs_score_texts(zs_pipe, texts, batch=BATCH):
    """Zero-shot scores for 'meeting intent' with truncation and robust fallback."""
    scores = np.zeros(len(texts), dtype=float)
    idxs = [i for i,t in enumerate(texts) if isinstance(t,str) and len(t)>3]
    if not idxs: return scores
    valid = [texts[i] for i in idxs]

    i = 0
    while i < len(valid):
        chunk = valid[i:i+batch]
        try:
            outs = zs_pipe(
                chunk, HYP_LABELS,
                hypothesis_template=HYP_TEMPLATE,
                multi_label=False,
                truncation=True
            )
        except Exception:
            outs = []
            for t in chunk:
                try:
                    o = zs_pipe(
                        t, HYP_LABELS,
                        hypothesis_template=HYP_TEMPLATE,
                        multi_label=False,
                        truncation=True
                    )
                    outs.append(o)
                except Exception:
                    outs.append({"labels":HYP_LABELS, "scores":[0.0,1.0]})
        # normalize outputs
        chunk_scores=[]
        for o in outs:
            if isinstance(o, list): o=o[0]
            label2score = {lbl: sc for lbl, sc in zip(o["labels"], o["scores"])}
            chunk_scores.append(float(label2score.get("meeting intent", 0.0)))
        scores[idxs[i:i+len(chunk_scores)]] = np.array(chunk_scores, dtype=float)
        i += batch
    return scores

# ---------------- LOAD & MERGE LABELS ----------------
assert SUB_COL in df_sample.columns and BODY_COL in df_sample.columns, f"df_sample must contain {SUB_COL}, {BODY_COL}"

lab = pd.read_csv(LABELS_CSV)
lab = lab.rename(columns={"subject":"subject_raw", "body":"body_train_v3"})  # normalize if needed
if LABEL_COL not in lab.columns:
    raise RuntimeError(f"Expected '{LABEL_COL}' in {LABELS_CSV}")
lab["__text"]  = combine_text(lab)
lab["__hash"]  = make_hash(lab["__text"])
lab["__label"] = lab[LABEL_COL].astype(bool)

df_all = df_sample.copy()
df_all["__text"] = combine_text(df_all)
df_all["__hash"] = make_hash(df_all["__text"])
df_all = df_all.reset_index().rename(columns={"index":"row_idx"})

# Merge by idx then by hash map (no shape mismatch)
merged = df_all.merge(lab[[IDX_COL, "__label"]].dropna(), left_on="row_idx", right_on=IDX_COL, how="left")
hash2label = (
    lab.dropna(subset=["__hash","__label"])
       .drop_duplicates("__hash")
       .set_index("__hash")["__label"]
)
merged["__label"] = merged["__label"].where(~merged["__label"].isna(), merged["__hash"].map(hash2label))
labeled = merged.dropna(subset=["__label"]).copy()
labeled["__label"] = labeled["__label"].astype(int)
print(f"Merged labels: {len(labeled)} rows (of ~{len(lab)})")

# ---------------- CALIBRATE THRESHOLD (cheap; small set) ----------------
X = labeled["__text"].tolist()
y = labeled["__label"].values
if len(np.unique(y)) < 2:
    raise RuntimeError("Your labeled set has only one class. Need both meeting and no-meeting examples.")

zs = pipeline("zero-shot-classification", model=MODEL_NAME, device_map="auto")
# Small val split for threshold selection
_, X_va, _, y_va = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
zs_va_scores = zs_score_texts(zs, X_va, batch=min(BATCH, 32))  # tiny set; smaller batch is fine
ap = average_precision_score(y_va, zs_va_scores)
prec, rec, thr = precision_recall_curve(y_va, zs_va_scores)
chosen_thr = 0.5
candidates = [(p, t) for p, t in zip(prec, np.r_[thr, 1.0]) if p >= TARGET_PREC]
if candidates:
    chosen_thr = max(t for p, t in candidates)
print(f"Validation AP: {ap:.4f}")
print(f"Chosen threshold for ≥{int(TARGET_PREC*100)}% precision: {chosen_thr:.3f}")
print("Validation report @ threshold:")
print(classification_report(y_va, (zs_va_scores >= chosen_thr).astype(int), target_names=["no_meeting","meeting"], digits=3))

# ---------------- FAST SHORTLIST (regex prefilter) ----------------
df_all["__text_low"] = df_all["__text"].str.lower()

# broad but high-recall keyword net
kw = r"(meet|meeting|schedule|scheduling|calendar|invite|invitation|call|zoom|teams|webex|google meet|hangout|catch up|sync|resched|appointment|room|conf\s*room)"
cand_mask = df_all["__text_low"].str.contains(kw, regex=True, na=False)

candidates = df_all[cand_mask].copy()
print(f"Candidates after regex: {len(candidates)} / {len(df_all)}")

# OPTIONAL: prioritize candidates that look time-like (quick regex) and cap to CANDIDATE_CAP
time_hint = r"(\bmon|tue|wed|thu|fri|sat|sun\b|\b\d{1,2}[:.]\d{2}\s*(am|pm)\b|\b\d{1,2}\s*(am|pm)\b|\b\d{4}-\d{2}-\d{2}\b|\bjan|feb|mar|apr|may|jun|jul|aug|sep|oct|nov|dec\b)"
candidates["__time_hit"] = candidates["__text_low"].str.contains(time_hint, regex=True, na=False).astype(int)

if len(candidates) > CANDIDATE_CAP:
    # take all with time hits first, then fill to cap arbitrarily
    with_time = candidates[candidates["__time_hit"] == 1]
    without_time = candidates[candidates["__time_hit"] == 0]
    head_with = with_time.head(int(CANDIDATE_CAP * 0.8))
    head_without = without_time.head(CANDIDATE_CAP - len(head_with))
    candidates = pd.concat([head_with, head_without], axis=0)
    print(f"Capped candidates to {len(candidates)} using time-hint priority.")

# ---------------- SCORE ONLY CANDIDATES ----------------
candidates = candidates.sort_values("row_idx")  # stable
cand_scores = zs_score_texts(zs, candidates["__text"].tolist(), batch=BATCH)
candidates["meeting_score"] = cand_scores

# Initialize all scores to 0.0, write back candidate scores
df_all["meeting_score"] = 0.0
df_all.loc[candidates.index, "meeting_score"] = candidates["meeting_score"].values

# ---------------- SELECT TOP-1000 POSITIVES ----------------
k = min(TOPK, len(df_all))
pos_topk_df = df_all.sort_values("meeting_score", ascending=False).head(k)

# ---------------- SAVE & EXPOSE ----------------
os.makedirs("artifacts", exist_ok=True)
export_cols = [c for c in ["from_email","to_emails","cc_emails",SUB_COL,BODY_COL,"meeting_score"] if c in pos_topk_df.columns]
if not export_cols:
    export_cols = ["row_idx","meeting_score","__text"]

pos_topk_df[export_cols].to_csv("artifacts/emails_meeting_POS_top1000.csv", index=False)
df_all.drop(columns=["__text_low","__time_hit"], errors="ignore").to_parquet("artifacts/scored_full.parquet", index=False)
labeled.to_parquet("artifacts/labels_merged.parquet", index=False)

print(f"\nSaved: artifacts/emails_meeting_POS_top1000.csv  (rows={len(pos_topk_df)})")
print(f"Saved: artifacts/scored_full.parquet             (rows={len(df_all)})")
print(f"Saved: artifacts/labels_merged.parquet           (rows={len(labeled)})")
```

```python
import re, json
import pandas as pd
from datetime import timedelta, datetime
import dateparser
from dateutil.parser import parse as dateutil_parse

# Paths
POS_CSV   = "emails_meeting_POS_top1000_dedup.csv"
OUT_JSONL = "finetune_calendar_events_FIXE.jsonl"

# Column names
SUB, BODY = "subject_raw", "body_train_v3"
FROM, TO, CC, BCC = "from_email", "to_emails", "cc_emails", "bcc_emails"
DATE_UTC = "date_utc"

def s(x):
    if x is None or (isinstance(x, float) and pd.isna(x)): return ""
    return x.strip() if isinstance(x, str) else str(x).strip()

def _emails(x):
    txt = s(x)
    if not txt: return []
    parts = re.split(r"[,\s;]+", txt.strip("[]"))
    out = []
    for p in parts:
        p = p.strip("'\"").lower()
        if re.match(r"[^@\s]+@[^@\s]+\.[^@\s]+", p or ""):
            out.append(p)
    return list(dict.fromkeys(out))

def _to_utc_iso(dt):
    if dt is None: return None
    ts = pd.Timestamp(dt)
    if getattr(ts, "tzinfo", None) is None or ts.tz is None:
        ts = ts.tz_localize("UTC")
    else:
        ts = ts.tz_convert("UTC")
    return ts.isoformat()

# === CRITICAL FIX: Better Time/Location Parser ===
def find_time_and_location(subject, body, email_sent_date_str):
    """
    FIXED: Parse subject and body SEPARATELY, prioritizing subject.
    Uses multiple strategies for robustness.
    """
    start_time, end_time, location = None, None, None
    notes_prefix = ""

    # --- 1. Set the Relative Base ---
    try:
        base_dt = dateutil_parse(email_sent_date_str)
    except:
        base_dt = datetime.utcnow()

    # --- 2. Priority Order: Subject first, then body ---
    # Subject usually has the most relevant meeting time
    search_texts = [
        ("subject", subject),
        ("body", body)
    ]

    for source, text in search_texts:
        if not text or start_time:  # Skip if already found
            continue

        # Strategy A: Look for explicit patterns first
        # Pattern: "Monday at 2pm", "Jan 15 at 10:30am", "tomorrow 3:00 PM"
        time_pattern = r'(?:(?:Monday|Tuesday|Wednesday|Thursday|Friday|Saturday|Sunday|Mon|Tue|Wed|Thu|Fri|Sat|Sun|tomorrow|today|next week)\s+(?:at\s+)?|(?:Jan(?:uary)?|Feb(?:ruary)?|Mar(?:ch)?|Apr(?:il)?|May|June?|July?|Aug(?:ust)?|Sep(?:tember)?|Oct(?:ober)?|Nov(?:ember)?|Dec(?:ember)?)\s+\d{1,2}(?:st|nd|rd|th)?\s+(?:at\s+)?|(?:\d{1,2}[/-]\d{1,2}(?:[/-]\d{2,4})?\s+(?:at\s+)?))(\d{1,2}(?::\d{2})?\s*(?:am|pm|AM|PM))'

        match = re.search(time_pattern, text, re.IGNORECASE)
        if match:
            # Extract the matched segment with context (20 chars before)
            match_start = max(0, match.start() - 20)
            context = text[match_start:match.end() + 10]
            parsed = dateparser.parse(
                context,
                settings={
                    "RELATIVE_BASE": base_dt,
                    "PREFER_DATES_FROM": "future",
                    "RETURN_AS_TIMEZONE_AWARE": True
                }
            )
            if parsed:
                start_time = parsed
                break

        # Strategy B: Look for time ranges "10am-11am", "2:00 to 3:30"
        range_pattern = r'(\d{1,2}(?::\d{2})?\s*(?:am|pm|AM|PM)?)\s*(?:to|-|–|until)\s*(\d{1,2}(?::\d{2})?\s*(?:am|pm|AM|PM))'
        range_match = re.search(range_pattern, text, re.IGNORECASE)
        if range_match:
            start_str = range_match.group(1)
            end_str = range_match.group(2)

            # Get surrounding context for better date parsing
            context_start = max(0, range_match.start() - 30)
            context_text = text[context_start:range_match.end() + 20]

            # Parse start time
            parsed_start = dateparser.parse(
                context_text,
                settings={
                    "RELATIVE_BASE": base_dt,
                    "PREFER_DATES_FROM": "future",
                    "RETURN_AS_TIMEZONE_AWARE": True
                }
            )

            if parsed_start:
                start_time = parsed_start

                # Parse end time relative to start date
                try:
                    # Extract just the time part for end time
                    end_time_only = dateparser.parse(
                        end_str,
                        settings={
                            "RELATIVE_BASE": start_time.replace(hour=0, minute=0),
                            "PREFER_DATES_FROM": "future",
                            "RETURN_AS_TIMEZONE_AWARE": True
                        }
                    )
                    if end_time_only:
                        # If end time is before start time (e.g., 10am-2pm), it's same day
                        if end_time_only.time() <= start_time.time():
                            end_time = start_time.replace(
                                hour=end_time_only.hour,
                                minute=end_time_only.minute
                            ) + timedelta(hours=12) if end_time_only.hour < 12 and start_time.hour >= 12 else start_time.replace(
                                hour=end_time_only.hour,
                                minute=end_time_only.minute
                            )
                        else:
                            end_time = end_time_only
                except:
                    end_time = start_time + timedelta(hours=1)
                break

        # Strategy C: Generic dateparser (last resort)
        if not start_time:
            # Try first 500 chars to avoid parsing random dates from body
            text_sample = text[:500] if source == "body" else text
            parsed = dateparser.parse(
                text_sample,
                settings={
                    "RELATIVE_BASE": base_dt,
                    "PREFER_DATES_FROM": "future",
                    "RETURN_AS_TIMEZONE_AWARE": True,
                    "STRICT_PARSING": False
                }
            )
            if parsed and parsed > base_dt - timedelta(days=1):  # Must be future or very recent
                start_time = parsed
                break

    # --- 3. Set end time if start found but no end ---
    if start_time and not end_time:
        end_time = start_time + timedelta(hours=1)

    if not start_time:
        notes_prefix = "Ambiguous or missing date/time. "

    # --- 4. Location Extraction (improved regex) ---
    combined_text = f"{subject} {body}"

    # Pattern priority:
    location_patterns = [
        # Room numbers and conference rooms
        r'\b(?:room|rm|conference room|meeting room)\s*[:#]?\s*([A-Z0-9\-]+(?:\s+[A-Z0-9\-]+)?)\b',
        # Virtual meeting platforms with links
        r'\b((?:zoom|teams|webex|google meet|meet\.google\.com)[^\s,]*(?:\s+meeting)?)\b',
        r'(https?://(?:zoom\.us|teams\.microsoft\.com|meet\.google\.com)[^\s]+)',
        # Building/floor locations
        r'\b(?:at|in|location:?)\s+([A-Z][a-z]+\s+(?:Building|Office|Center|Hall)(?:,?\s+(?:Floor|Room)\s+[A-Z0-9\-]+)?)\b',
        # Generic location indicator
        r'\blocation:?\s+([A-Za-z0-9\s,\-]+?)(?:\.|,|\n|$)',
    ]

    for pattern in location_patterns:
        match = re.search(pattern, combined_text, re.IGNORECASE)
        if match:
            location = match.group(1).strip()
            # Clean up common issues
            location = re.sub(r'\s+', ' ', location)  # Normalize whitespace
            location = location.rstrip('.,;:')  # Remove trailing punctuation
            break

    return start_time, end_time, location, notes_prefix


def draft_event_FIXED(row):
    """FIXED: Uses improved parsing that separates subject and body"""
    subj = s(row.get(SUB, ""))
    body = s(row.get(BODY, ""))
    email_date = s(row.get(DATE_UTC, ""))

    # CRITICAL: Parse subject and body separately
    start, end, loc, notes_prefix = find_time_and_location(subj, body, email_date)

    attendees = list(dict.fromkeys(
        _emails(row.get(TO, "")) +
        _emails(row.get(CC, "")) +
        _emails(row.get(FROM, "")) +
        _emails(row.get(BCC, ""))
    ))

    return {
        "title": (subj or "Meeting")[:200],
        "start_utc": _to_utc_iso(start),
        "end_utc": _to_utc_iso(end),
        "location": loc,
        "attendees": attendees[:50],
        "notes": (notes_prefix + body)[:1000]
    }

# === FEW-SHOT EXAMPLES: Teaching the model date/time extraction ===
def create_few_shot_examples(base_datetime):
    """
    Creates synthetic examples that explicitly teach date/time extraction patterns.
    These are prepended to the training data.
    """
    examples = []

    # Example 1: Simple time with relative date
    examples.append({
        "instruction": (
            "You are a precise calendar event extraction assistant. "
            "Extract the calendar event. Return ONLY a JSON object with keys: "
            "title, start_utc, end_utc, location, attendees, notes.\n"
            "- All timestamps must be ISO8601 with timezone or Z.\n"
            "- If time is ambiguous, set start_utc and end_utc to null and explain in notes.\n"
            "- Attendees should include emails from From, To, and Cc."
        ),
        "input": """Subject: Team Sync - Monday at 2pm
From: alice@company.com
To: bob@company.com
Cc: charlie@company.com
Bcc:

Body:
Hey team, let's meet to discuss Q1 progress.""",
        "output": {
            "title": "Team Sync - Monday at 2pm",
            "start_utc": "2025-01-06T14:00:00+00:00",  # Next Monday 2pm UTC
            "end_utc": "2025-01-06T15:00:00+00:00",
            "location": None,
            "attendees": ["alice@company.com", "bob@company.com", "charlie@company.com"],
            "notes": "Hey team, let's meet to discuss Q1 progress."
        }
    })

    # Example 2: Time range with location
    examples.append({
        "instruction": (
            "You are a precise calendar event extraction assistant. "
            "Extract the calendar event. Return ONLY a JSON object with keys: "
            "title, start_utc, end_utc, location, attendees, notes.\n"
            "- All timestamps must be ISO8601 with timezone or Z.\n"
            "- If time is ambiguous, set start_utc and end_utc to null and explain in notes.\n"
            "- Attendees should include emails from From, To, and Cc."
        ),
        "input": """Subject: Budget Review Meeting
From: manager@company.com
To: team@company.com
Cc:
Bcc:

Body:
Meeting tomorrow from 10:00 AM to 11:30 AM in Conference Room B.
Please bring your reports.""",
        "output": {
            "title": "Budget Review Meeting",
            "start_utc": "2025-01-04T10:00:00+00:00",  # Tomorrow 10am UTC
            "end_utc": "2025-01-04T11:30:00+00:00",
            "location": "Conference Room B",
            "attendees": ["manager@company.com", "team@company.com"],
            "notes": "Meeting tomorrow from 10:00 AM to 11:30 AM in Conference Room B.\nPlease bring your reports."
        }
    })

    # Example 3: Specific date and time
    examples.append({
        "instruction": (
            "You are a precise calendar event extraction assistant. "
            "Extract the calendar event. Return ONLY a JSON object with keys: "
            "title, start_utc, end_utc, location, attendees, notes.\n"
            "- All timestamps must be ISO8601 with timezone or Z.\n"
            "- If time is ambiguous, set start_utc and end_utc to null and explain in notes.\n"
            "- Attendees should include emails from From, To, and Cc."
        ),
        "input": """Subject: Product Launch Planning
From: product@company.com
To: engineering@company.com, marketing@company.com
Cc: ceo@company.com
Bcc:

Body:
Let's meet on January 15th at 3:00 PM to finalize the product launch timeline.
Location: Main Office, Building A, Room 405.
This is a critical meeting for Q1 goals.""",
        "output": {
            "title": "Product Launch Planning",
            "start_utc": "2025-01-15T15:00:00+00:00",
            "end_utc": "2025-01-15T16:00:00+00:00",
            "location": "Main Office, Building A, Room 405",
            "attendees": ["product@company.com", "engineering@company.com", "marketing@company.com", "ceo@company.com"],
            "notes": "Let's meet on January 15th at 3:00 PM to finalize the product launch timeline.\nLocation: Main Office, Building A, Room 405.\nThis is a critical meeting for Q1 goals."
        }
    })

    # Example 4: Virtual meeting
    examples.append({
        "instruction": (
            "You are a precise calendar event extraction assistant. "
            "Extract the calendar event. Return ONLY a JSON object with keys: "
            "title, start_utc, end_utc, location, attendees, notes.\n"
            "- All timestamps must be ISO8601 with timezone or Z.\n"
            "- If time is ambiguous, set start_utc and end_utc to null and explain in notes.\n"
            "- Attendees should include emails from From, To, and Cc."
        ),
        "input": """Subject: Weekly Standup - Friday 9am
From: scrum.master@company.com
To: dev.team@company.com
Cc:
Bcc:

Body:
Join us this Friday at 9:00 AM for our weekly standup.
Zoom link: https://zoom.us/j/123456789
Duration: 30 minutes""",
        "output": {
            "title": "Weekly Standup - Friday 9am",
            "start_utc": "2025-01-10T09:00:00+00:00",
            "end_utc": "2025-01-10T09:30:00+00:00",
            "location": "https://zoom.us/j/123456789",
            "attendees": ["scrum.master@company.com", "dev.team@company.com"],
            "notes": "Join us this Friday at 9:00 AM for our weekly standup.\nZoom link: https://zoom.us/j/123456789\nDuration: 30 minutes"
        }
    })

    # Example 5: Ambiguous time (teaching null handling)
    examples.append({
        "instruction": (
            "You are a precise calendar event extraction assistant. "
            "Extract the calendar event. Return ONLY a JSON object with keys: "
            "title, start_utc, end_utc, location, attendees, notes.\n"
            "- All timestamps must be ISO8601 with timezone or Z.\n"
            "- If time is ambiguous, set start_utc and end_utc to null and explain in notes.\n"
            "- Attendees should include emails from From, To, and Cc."
        ),
        "input": """Subject: Catch up soon?
From: colleague@company.com
To: me@company.com
Cc:
Bcc:

Body:
Hey, we should grab coffee sometime next week and discuss the project.
Let me know what works for you!""",
        "output": {
            "title": "Catch up soon?",
            "start_utc": None,
            "end_utc": None,
            "location": None,
            "attendees": ["colleague@company.com", "me@company.com"],
            "notes": "Ambiguous or missing date/time. Hey, we should grab coffee sometime next week and discuss the project.\nLet me know what works for you!"
        }
    })

    return examples

# === MAIN DATA GENERATION ===
print(f"Reading {POS_CSV}...")
df = pd.read_csv(POS_CSV)

if BCC not in df.columns:
    df[BCC] = ""

# Generate few-shot examples first
base_dt = datetime(2025, 1, 3, 12, 0, 0)  # Friday, Jan 3, 2025 12pm UTC
few_shot_examples = create_few_shot_examples(base_dt)
print(f"Created {len(few_shot_examples)} few-shot teaching examples")

pairs = []
parse_successes = 0

# Process real data
for _, r in df.iterrows():
    subj_raw = s(r.get(SUB, ""))
    body_raw = s(r.get(BODY, ""))

    email_text = (
        "Subject: " + subj_raw + "\n"
        "From: "    + s(r.get(FROM, "")) + "\n"
        "To: "      + s(r.get(TO, ""))   + "\n"
        "Cc: "      + s(r.get(CC, ""))   + "\n"
        "Bcc: "     + s(r.get(BCC, ""))  + "\n\n"
        "Body:\n"   + body_raw
    ).strip()

    out = draft_event_FIXED(r)

    if out['start_utc'] is not None:
        parse_successes += 1

    pairs.append({
        "instruction": (
            "You are a precise calendar event extraction assistant. "
            "Extract the calendar event. Return ONLY a JSON object with keys: "
            "title, start_utc, end_utc, location, attendees, notes.\n"
            "- All timestamps must be ISO8601 with timezone or Z.\n"
            "- If time is ambiguous, set start_utc and end_utc to null and explain in notes.\n"
            "- Attendees should include emails from From, To, and Cc."
        ),
        "input": email_text,
        "output": out
    })

# CRITICAL: Prepend few-shot examples so model sees them first
all_pairs = few_shot_examples + pairs

with open(OUT_JSONL, "w", encoding="utf-8") as f:
    for p in all_pairs:
        f.write(json.dumps(p, ensure_ascii=False) + "\n")

print(f"\n{'='*60}")
print(f"✅ Wrote {len(all_pairs)} total pairs -> {OUT_JSONL}")
print(f"   - {len(few_shot_examples)} few-shot teaching examples")
print(f"   - {len(pairs)} real email examples")
print(f"{'='*60}")
print(f"📊 PARSING STATS:")
print(f"   Successfully parsed dates: {parse_successes}/{len(pairs)} ({100*parse_successes/len(pairs):.1f}%)")
print(f"{'='*60}")
```

```python
import re, json
import pandas as pd
from datetime import timedelta
import dateparser
from dateutil.parser import parse as dateutil_parse

# Paths
POS_CSV   = "emails_meeting_POS_top1000_dedup.csv"
OUT_JSONL = "finetune_calendar_events_FIXED.jsonl" # New output file

# Column names
SUB, BODY = "subject_raw", "body_train_v3"
FROM, TO, CC, BCC = "from_email", "to_emails", "cc_emails", "bcc_emails"
DATE_UTC = "date_utc" # The most important column!

def s(x):
    if x is None or (isinstance(x, float) and pd.isna(x)): return ""
    return x.strip() if isinstance(x, str) else str(x).strip()

def _emails(x):
    txt = s(x)
    if not txt: return []
    parts = re.split(r"[,\s;]+", txt.strip("[]"))
    out = []
    for p in parts:
        p = p.strip("'\"").lower()
        if re.match(r"[^@\s]+@[^@\s]+\.[^@\s]+", p or ""):
            out.append(p)
    return list(dict.fromkeys(out))

def _to_utc_iso(dt):
    if dt is None: return None
    ts = pd.Timestamp(dt)
    if getattr(ts, "tzinfo", None) is None or ts.tz is None:
        ts = ts.tz_localize("UTC")
    else:
        ts = ts.tz_convert("UTC")
    return ts.isoformat()

# --- THIS IS THE NEW, SMARTER PARSING LOGIC ---

def find_time_and_location(text_to_parse, email_sent_date_str):
    """
    Smarter parser for time and location.
    """
    start_time, end_time, location = None, None, None
    notes_prefix = ""

    # --- 1. Set the Relative Base Correctly ---
    # Use the email's sent date as the base, not "now"
    try:
        base_dt = dateutil_parse(email_sent_date_str)
    except:
        base_dt = dateparser.parse("now UTC") # Fallback if email date is bad

    # --- 2. Try to find a start time ---
    parsed_start = dateparser.parse(text_to_parse, settings={"RELATIVE_BASE": base_dt})

    if parsed_start:
        start_time = parsed_start

        # --- 3. Try to find an end time ---
        # Regex for patterns like "10am to 11:30am" or "10-11:30"
        # This is complex, so we'll try a simpler 'duration' parse
        # (A more robust solution would use a dedicated library)

        # Look for "from [time] to [time]"
        end_time_match = re.search(
            r"(?:from|between)\s+(.*?)\s+(?:to|and|-)\s+(.*?)(?:\s+|$|\n|,|\.)",
            text_to_parse, re.I
        )
        if end_time_match:
            try:
                # Try to parse the end time string relative to the *start date*
                end_time_str = end_time_match.group(2)
                end_time = dateparser.parse(end_time_str, settings={"RELATIVE_BASE": start_time.date()})

                # Handle cases like "10 AM to 2 PM" - if end is before start, add 12 hours (PM)
                if end_time and end_time < start_time:
                     end_time += timedelta(hours=12)
            except:
                pass # Stick with default

        # Default: if no end time found, set to 1 hour after start
        if not end_time:
            end_time = start_time + timedelta(minutes=60)

    else:
        notes_prefix = "Ambiguous or missing date/time. "

    # --- 4. Try to find location ---
    # A better regex that matches "room [name]"
    loc_match = re.search(
        r"\b(?:in|at|room|location)\s+((?:[A-Z0-9\-]+|the)\s*(?:conference|study|meeting)?\s*(?:room)?\s*[A-Z0-9\-#B]+)\b",
        text_to_parse, re.I
    )
    if loc_match:
        location = loc_match.group(1).strip()

    return start_time, end_time, location, notes_prefix


def draft_event_FIXED(row):
    subj = s(row.get(SUB, ""))
    body = s(row.get(BODY, ""))
    txt  = f"{subj}\n\n{body}".strip()

    # ** THE FIX: Pass the email's date to the parser **
    email_date = s(row.get(DATE_UTC, ""))
    start, end, loc, notes_prefix = find_time_and_location(txt, email_date)

    attendees = list(dict.fromkeys(
        _emails(row.get(TO, "")) + _emails(row.get(CC, "")) + _emails(row.get(FROM, "")) + _emails(row.get(BCC, ""))
    ))

    return {
        "title": (subj or "Meeting")[:200],
        "start_utc": _to_utc_iso(start),
        "end_utc": _to_utc_iso(end),
        "location": loc,
        "attendees": attendees[:50],
        "notes": (notes_prefix + body)[:1000]
    }

# ---- build pairs ----
print(f"Reading {POS_CSV}...")
df = pd.read_csv(POS_CSV)

if BCC not in df.columns:
    df[BCC] = ""

pairs = []
parse_successes = 0
for _, r in df.iterrows():
    subj_raw = s(r.get(SUB, ""))
    body_raw = s(r.get(BODY, ""))

    # Construct the *input* text
    email_text = (
        "Subject: " + subj_raw + "\n"
        "From: "    + s(r.get(FROM, "")) + "\n"
        "To: "      + s(r.get(TO, ""))   + "\n"
        "Cc: "      + s(r.get(CC, ""))   + "\n"
        "Bcc: "     + s(r.get(BCC, ""))  + "\n\n"
        "Body:\n"   + body_raw
    ).strip()

    # Construct the *output* JSON using the NEW function
    out = draft_event_FIXED(r)

    if out['start_utc'] is not None:
        parse_successes += 1

    pairs.append({
        "instruction": (
            "You are a precise calendar event extraction assistant. "
            "Extract the calendar event. Return ONLY a JSON object with keys: "
            "title, start_utc, end_utc, location, attendees, notes.\n"
            "- All timestamps must be ISO8601 with timezone or Z.\n"
            "- If time is ambiguous, set start_utc and end_utc to null and explain in notes.\n"
            "- Attendees should include emails from From, To, and Cc."
        ),
        "input": email_text,
        "output": out
    })

with open(OUT_JSONL, "w", encoding="utf-8") as f:
    for p in pairs:
        f.write(json.dumps(p, ensure_ascii=False) + "\n")

print(f"Wrote draft pairs: {len(pairs)} -> {OUT_JSONL}")
print(f"--- PARSING STATS ---")
print(f"Successfully parsed a non-null date for {parse_successes} / {len(pairs)} emails.")
print(f"This is MUCH better than the 0 successes from the original script.")
```

```python
from transformers import TrainingArguments, Trainer
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
from datasets import Dataset
import torch
import numpy as np

print("="*80)
print("STEP 1: Prepare Model for Training")
print("="*80)

model = prepare_model_for_kbit_training(model)

# LoRA config for Mistral (q_proj, v_proj are common)
lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=[
        "q_proj", "k_proj", "v_proj", "o_proj",
        "gate_proj", "up_proj", "down_proj"
    ],
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM"
)

model = get_peft_model(model, lora_config)
model.print_trainable_parameters()

print("\n" + "="*80)
print("STEP 2: Tokenize Training Data")
print("="*80)

def tokenize_texts(texts, tokenizer, max_length=2048):
    """Tokenize list of formatted strings"""
    # Note: We tokenize without padding. The collator will handle it.
    tokenized_outputs = tokenizer(
        texts,
        truncation=True,
        max_length=max_length,
        padding=False,
        return_tensors=None,
    )
    # We only need input_ids. Labels will be created by the collator.
    return {'input_ids': tokenized_outputs['input_ids']}

print(f"Tokenizing {len(train_texts)} training examples...")
train_tokenized = tokenize_texts(train_texts, tok)
print(f"Tokenizing {len(val_texts)} validation examples...")
val_tokenized = tokenize_texts(val_texts, tok)

train_dataset = Dataset.from_dict(train_tokenized)
val_dataset = Dataset.from_dict(val_tokenized)
print(f"✅ Train dataset: {len(train_dataset)}, Val dataset: {len(val_dataset)}")


print("\n" + "="*80)
print("STEP 3: Configure Training")
print("="*80)

training_args = TrainingArguments(
    output_dir="./calendar_extraction_model",
    num_train_epochs=3,
    per_device_train_batch_size=4,   # Reduce to 2 or 1 if OOM
    per_device_eval_batch_size=4,
    gradient_accumulation_steps=4,   # Effective batch size = 16
    learning_rate=2e-4,
    weight_decay=0.01,
    warmup_ratio=0.1,
    lr_scheduler_type="cosine",
    optim="paged_adamw_8bit",
    fp16=False,
    bf16=torch.cuda.is_bf16_supported(),
    logging_steps=10,
    eval_strategy="steps",
    eval_steps=50,
    save_strategy="steps",
    save_steps=50,
    save_total_limit=3,
    load_best_model_at_end=True,
    metric_for_best_model="eval_loss",
    greater_is_better=False,
    report_to="none",
    remove_unused_columns=False, # Important for our custom collator
)

print("\n" + "="*80)
print("STEP 4: Initialize Trainer with *Advanced* Data Collator")
print("="*80)

class DataCollatorForMaskedLM:
    """
    **ADVANCED COLLATOR**
    Pads sequences and masks prompt tokens (everything up to and
    including '[/INST]') in the labels.
    """
    def __init__(self, tokenizer):
        self.tokenizer = tokenizer
        # Tokenize the end-of-instruction tag to find it later
        # We use add_special_tokens=False to get just the raw token IDs
        self.inst_end_tokens = self.tokenizer.encode("[/INST]", add_special_tokens=False)
        self.inst_end_len = len(self.inst_end_tokens)
        print(f"Collator: Found [/INST] tokens: {self.inst_end_tokens}")

    def __call__(self, features):
        batch_input_ids = [f['input_ids'] for f in features]

        # 1. Find max length in batch
        max_length = max(len(ids) for ids in batch_input_ids)

        # 2. Prepare padded tensors
        padded_input_ids = []
        padded_labels = []
        padded_attention_mask = []

        for input_ids in batch_input_ids:
            seq_len = len(input_ids)
            padding_len = max_length - seq_len

            # 3. Find the [/INST] boundary
            # We search for the token sequence of "[/INST]"
            split_point = -1
            for i in range(seq_len - self.inst_end_len + 1):
                if input_ids[i:i+self.inst_end_len] == self.inst_end_tokens:
                    split_point = i + self.inst_end_len
                    break

            # 4. Create labels: -100 for prompt, copy of input_ids for response
            if split_point != -1:
                prompt_len = split_point
                response_len = seq_len - prompt_len

                # Mask prompt tokens
                labels = [-100] * prompt_len
                # Keep response tokens
                labels.extend(input_ids[prompt_len:])
                # Mask padding tokens
                labels.extend([-100] * padding_len)
            else:
                # Fallback: if [/INST] not found (shouldn't happen), mask nothing
                print("Warning: [/INST] not found in sequence. Training on full sequence.")
                labels = list(input_ids) + [-100] * padding_len

            # 5. Pad input_ids and attention_mask
            padded_ids = input_ids + [self.tokenizer.pad_token_id] * padding_len
            padded_mask = [1] * seq_len + [0] * padding_len

            padded_input_ids.append(padded_ids)
            padded_attention_mask.append(padded_mask)
            padded_labels.append(labels)

        # 6. Convert to tensors
        return {
            'input_ids': torch.tensor(padded_input_ids, dtype=torch.long),
            'attention_mask': torch.tensor(padded_attention_mask, dtype=torch.long),
            'labels': torch.tensor(padded_labels, dtype=torch.long),
        }

data_collator = DataCollatorForMaskedLM(tokenizer=tok)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=tok,
    data_collator=data_collator,
)

print("✅ Trainer initialized with *masking* data collator")

print("\n" + "="*80)
print("STEP 5: Start Training")
print("="*80)

try:
    trainer.train()
    print("\n🎉 Training completed successfully!")
except Exception as e:
    print(f"\n❌ Training failed with error: {e}")
    import traceback
    traceback.print_exc()
    raise

print("\n" + "="*80)
print("STEP 6: Save Final Model")
print("="*80)

final_path = "./calendar_extraction_model_final"
trainer.save_model(final_path)
tok.save_pretrained(final_path)

print(f"✅ Final model (LoRA adapters) saved to: {final_path}")
```

```python
import json
import numpy as np
from collections import defaultdict
from sklearn.metrics import precision_recall_fscore_support, confusion_matrix

def read_jsonl(path):
    """Reads a .jsonl file and returns a list of dictionaries."""
    with open(path, "r", encoding="utf-8") as f:
        return [json.loads(l) for l in f]

def calculate_set_f1(pred_list, truth_list):
    """
    Calculates Precision, Recall, and F1 for a list of items (like attendees).
    Order does not matter.
    """
    # Handle Nones or empty lists
    pred_set = set(e.lower() for e in pred_list if e)
    truth_set = set(e.lower() for e in truth_list if e)

    if not truth_set and not pred_set:
        return 1.0, 1.0, 1.0  # Correctly predicted nothing

    if not truth_set:
        return 0.0, 1.0, 0.0 # Predicted items when none were expected (perfect recall, 0 precision)

    if not pred_set:
        return 1.0, 0.0, 0.0 # Predicted nothing when items were expected (perfect precision, 0 recall)

    intersection = len(pred_set.intersection(truth_set))

    precision = intersection / len(pred_set)
    recall = intersection / len(truth_set)

    if (precision + recall) == 0:
        f1 = 0.0
    else:
        f1 = 2 * (precision * recall) / (precision + recall)

    return precision, recall, f1

def evaluate_predictions(file_path="test_set_predictions.jsonl"):
    """
    Reads the prediction file and computes a wide range of metrics.
    """
    print(f"Starting evaluation of: {file_path}")

    try:
        results = read_jsonl(file_path)
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
        print("Please run the batch inference script first.")
        return

    total_samples = len(results)
    if total_samples == 0:
        print("Error: No results found in file.")
        return

    # --- 1. Overall Health Counters ---
    json_valid_count = 0
    schema_compliant_count = 0
    full_exact_match_count = 0

    # List to store raw outputs for failed JSON parsing
    json_failure_examples = []

    # --- 2. Field-Level Aggregators ---
    field_em = defaultdict(list)
    attendee_scores = {'precision': [], 'recall': [], 'f1': []}

    # For timestamp null/not-null confusion matrix
    # 0 = null (ambiguous), 1 = not-null (time found)
    y_true_time = []
    y_pred_time = []

    # For ambiguity note accuracy
    ambiguity_note_correct = []

    REQUIRED_KEYS = {'title', 'start_utc', 'end_utc', 'location', 'attendees', 'notes'}

    for i, res in enumerate(results):
        pred_json = res.get('predicted_json')
        truth_json = res.get('ground_truth_json')

        # --- Metric 1: JSON Validity ---
        if pred_json is None:
            json_failure_examples.append(res.get('raw_model_output', ''))
            continue
        json_valid_count += 1

        # --- Metric 2: Schema Compliance ---
        pred_keys = set(pred_json.keys())
        if not REQUIRED_KEYS.issubset(pred_keys):
            # print(f"Warning: Sample {i} missing keys. Has: {pred_keys}")
            continue
        schema_compliant_count += 1

        # --- From here on, we have valid, compliant JSON ---

        # --- Metric 3: Overall Exact Match ---
        if pred_json == truth_json:
            full_exact_match_count += 1

        # --- Metric 4: Simple Field EM (title, location) ---
        field_em['title'].append(pred_json['title'] == truth_json['title'])
        field_em['location'].append(pred_json['location'] == truth_json['location'])

        # --- Metric 5: Attendee F1-Score ---
        p, r, f1 = calculate_set_f1(pred_json['attendees'], truth_json['attendees'])
        attendee_scores['precision'].append(p)
        attendee_scores['recall'].append(r)
        attendee_scores['f1'].append(f1)

        # --- Metric 6: Timestamp Null/Not-Null ---
        truth_has_time = 1 if truth_json['start_utc'] is not None else 0
        pred_has_time = 1 if pred_json['start_utc'] is not None else 0
        y_true_time.append(truth_has_time)
        y_pred_time.append(pred_has_time)

        # Also check exact match *if* time was supposed to be present
        if truth_has_time:
            field_em['start_utc'].append(pred_json['start_utc'] == truth_json['start_utc'])
            field_em['end_utc'].append(pred_json['end_utc'] == truth_json['end_utc'])

        # --- Metric 7: Ambiguity Note Accuracy ---
        truth_is_ambiguous = truth_json['start_utc'] is None
        pred_has_warning = str(pred_json['notes']).startswith("Ambiguous")
        ambiguity_note_correct.append(truth_is_ambiguous == pred_has_warning)


    # --- 3. Calculate and Print Report ---
    print("\n" + "="*80)
    print(" 📊 EVALUATION REPORT")
    print("="*80)

    print(f"--- 1. Overall Model Health ({total_samples} samples) ---")
    print(f"JSON Validity Rate:     {json_valid_count / total_samples:.2%} ({json_valid_count}/{total_samples})")
    valid_pct_base = json_valid_count if json_valid_count > 0 else 1
    print(f"Schema Compliance Rate: {schema_compliant_count / valid_pct_base:.2%} ({schema_compliant_count}/{json_valid_count})")
    schema_pct_base = schema_compliant_count if schema_compliant_count > 0 else 1
    print(f"Overall Exact Match:    {full_exact_match_count / schema_pct_base:.2%} ({full_exact_match_count}/{schema_compliant_count})")

    print("\n--- 2. Field-Level Exact Match (EM) Accuracy ---")
    print(f" (Based on {schema_compliant_count} schema-compliant samples)")
    print(f"Title:                  {np.mean(field_em['title']):.2%}")
    print(f"Location:               {np.mean(field_em['location']):.2%}")
    print(f"Ambiguity Note Logic:   {np.mean(ambiguity_note_correct):.2%}")
    print(f"Start UTC (when found): {np.mean(field_em['start_utc']):.2%} (from {len(field_em['start_utc'])} non-null samples)")
    print(f"End UTC (when found):   {np.mean(field_em['end_utc']):.2%} (from {len(field_em['end_utc'])} non-null samples)")

    print("\n--- 3. Attendee Performance (Your Key Metric) ---")
    print(f" (Based on {schema_compliant_count} schema-compliant samples)")
    print(f"Average Precision:      {np.mean(attendee_scores['precision']):.2%}")
    print(f"Average Recall:         {np.mean(attendee_scores['recall']):.2%}")
    print(f"Average F1-Score:       {np.mean(attendee_scores['f1']):.2%}")

    print("\n--- 4. Timestamp Ambiguity Analysis (Confusion Matrix) ---")
    if y_true_time:
        cm = confusion_matrix(y_true_time, y_pred_time, labels=[1, 0])
        print("          PREDICTED:")
        print("          Time    |  Null (Ambiguous)")
        print("         -------------------------------")
        print(f"ACTUAL | Time   |  {cm[0][0]:<6} |  {cm[0][1]:<6}  (Correctly Found | Missed Time)")
        print(f"ACTUAL | Null   |  {cm[1][0]:<6} |  {cm[1][1]:<6}  (Hallucinated Time | Correctly Null)")
        #
    else:
        print("No valid samples to build a confusion matrix.")

    print("\n" + "="*80)
    print("--- 5. Failure Analysis: Invalid JSON Outputs ---")
    print("="*80)
    if json_failure_examples:
        print(f"Found {len(json_failure_examples)} raw outputs that failed JSON parsing.")
        print("Showing first 5 examples:")
        for i, example in enumerate(json_failure_examples[:5]):
            print(f"\nFAILURE {i+1}:\n---\n{example}\n---")
    else:
        print("✅ No JSON parsing failures. Excellent!")

    print("\n" + "="*80)
    print("Evaluation Complete.")

# --- Run the evaluation ---
if __name__ == "__main__":
    evaluate_predictions("test_set_predictions.jsonl")
```

```python
import json
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from peft import PeftModel
import icalendar
from datetime import datetime
import dateutil.parser
import sys
import os

# --- 1. Helper Function: Load Model ---
def load_trained_model(base_model_id, adapter_path):
    """Loads the base model and applies the LoRA adapters."""

    if not os.path.exists(adapter_path):
        print(f"Error: Adapter path not found: {adapter_path}", file=sys.stderr)
        print("Please ensure the model was trained and saved to this location.", file=sys.stderr)
        return None, None

    print(f"Loading base model: {base_model_id}...")
    try:
        bnb = BitsAndBytesConfig(
            load_in_4bit=True,
            bnb_4bit_quant_type="nf4",
            bnb_4bit_compute_dtype=torch.bfloat16,
        )

        tokenizer = AutoTokenizer.from_pretrained(base_model_id)
        if tokenizer.pad_token is None:
            tokenizer.pad_token = tokenizer.eos_token
        tokenizer.padding_side = "right"

        base_model = AutoModelForCausalLM.from_pretrained(
            base_model_id,
            quantization_config=bnb,
            device_map="auto",
        )

        print(f"Applying LoRA adapters from: {adapter_path}...")
        model = PeftModel.from_pretrained(base_model, adapter_path)
        model = model.eval()
        print("✅ Model loaded and ready.")
        return model, tokenizer
    except Exception as e:
        print(f"Error loading model: {e}", file=sys.stderr)
        import traceback
        traceback.print_exc()
        return None, None

# --- 2. Helper Function: Prompt & Inference ---
def format_inference_prompt(email_text):
    """Formats the raw email text into the Mistral prompt template."""
    instruction = (
        "You are a precise calendar event extraction assistant. "
        "Extract the calendar event. Return ONLY a JSON object with keys: "
        "title, start_utc, end_utc, location, attendees, notes.\n"
        "- All timestamps must be ISO8601 with timezone or Z.\n"
        "- If time is ambiguous, set start_utc and end_utc to null and explain in notes.\n"
        "- Attendees should include emails from From, To, and Cc."
    )
    prompt = f"[INST] {instruction.strip()}\n\nHere is the email:\n\n{email_text.strip()} [/INST]"
    return prompt

def run_inference(model, tokenizer, email_text, max_new_tokens=1024):
    """Runs inference and attempts to parse the JSON output."""
    prompt = format_inference_prompt(email_text)
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            pad_token_id=tokenizer.eos_token_id,
            do_sample=False
        )

    response_text = tokenizer.decode(outputs[0][inputs['input_ids'].shape[1]:], skip_special_tokens=True)

    try:
        json_start = response_text.find('{')
        json_end = response_text.rfind('}') + 1
        if json_start != -1 and json_end != -1:
            json_str = response_text[json_start:json_end]
            return json.loads(json_str), response_text
        else:
            return None, response_text
    except json.JSONDecodeError:
        print(f"\nWarning: Failed to parse JSON from output:\n{response_text}\n", file=sys.stderr)
        return None, response_text

# --- 3. Helper Function: ICS Conversion ---
def convert_json_to_ics(event_json):
    """Converts the model's JSON output to an .ics file string."""
    cal = icalendar.Calendar()
    cal.add('prodid', '-//Calendar Extraction Model//')
    cal.add('version', '2.0')
    event = icalendar.Event()

    event.add('summary', event_json.get('title', 'No Title'))
    event.add('description', event_json.get('notes', ''))

    loc = event_json.get('location')
    if loc:
        event.add('location', loc)

    try:
        start_utc = event_json.get('start_utc')
        end_utc = event_json.get('end_utc')

        if start_utc and end_utc:
            dt_start = dateutil.parser.isoparse(start_utc)
            dt_end = dateutil.parser.isoparse(end_utc)
            event.add('dtstart', dt_start)
            event.add('dtend', dt_end)
        elif start_utc: # Handle if only start_utc is provided
            dt_start = dateutil.parser.isoparse(start_utc)
            event.add('dtstart', dt_start)
            event.add('dtend', dt_start + datetime.timedelta(hours=1)) # Default 1 hour
        else:
            event.add('dtstart', datetime.now().date())

    except Exception as e:
        print(f"Error parsing datetime: {e}. Skipping times.", file=sys.stderr)
        event.add('dtstart', datetime.now().date()) # Fallback

    attendees = event_json.get('attendees', [])
    for email in attendees:
        attendee = icalendar.vCalAddress(f'MAILTO:{email}')
        event.add('attendee', attendee, encode=0)

    cal.add_component(event)
    return cal.to_ical().decode('utf-8')

# --- 4. Main Execution ---
def main_test():
    BASE_MODEL = "mistralai/Mistral-7B-Instruct-v0.2"
    ADAPTER_PATH = "./calendar_extraction_model_final"

    # Load the model
    model, tokenizer = load_trained_model(BASE_MODEL, ADAPTER_PATH)

    if model is None or tokenizer is None:
        print("Exiting due to model loading failure.")
        return

    # ** New dummy email with start and end time **
    # Note: "tomorrow" will be interpreted relative to your system's current date.
    dummy_email_text = (
        "Subject: Q4 Planning Session\n"
        "From: ketki.maddiwar@sjsu.edu\n"
        "To: leelaprasad.dammalapati@sjsu.edu\n"
        "Cc: utkarsh.tripathi@sjsu.edu, aishanee.sinha@sjsu.edu\n"
        "Bcc: \n\n"
        "Body:\n"
        "Hi all,\n\n"
        "Following up on our last discussion, let's schedule the Q4 planning session.\n\n"
        "Does tomorrow, November 4th, from 10:00 AM to 11:30 AM work for everyone? We can use the main conference room.\n\n"
        "Please confirm.\n\n"
        "Best,\n"
        "Ketki"
    ).strip()

    print("="*80)
    print(f"--- INPUT EMAIL ---\n{dummy_email_text}\n{'-'*20}")

    # Run inference
    predicted_json, raw_output = run_inference(model, tokenizer, dummy_email_text)

    if predicted_json:
        print(f"--- PARSED JSON ---\n{json.dumps(predicted_json, indent=2)}\n{'-'*20}")

        # Convert to ICS
        ics_string = convert_json_to_ics(predicted_json)

        print(f"--- GENERATED .ICS FILE ---\n{ics_string}\n{'-'*20}")

        # Save to file
        output_filename = "dummy_meeting_with_end_time.ics"
        with open(output_filename, "w") as f:
            f.write(ics_string)
        print(f"✅ Successfully saved to {output_filename}")
        print("You can now import this file into your calendar.")

    else:
        print("❌ Model did not produce valid JSON.")
        print(f"--- RAW MODEL OUTPUT ---\n{raw_output}")

# --- This makes the script runnable ---
if __name__ == "__main__":
    main_test()
```
