```python
from transformers import pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
long_email = """
Subject: Project Update and Next Steps

Dear Team,

I hope this email finds you all well. I wanted to provide a comprehensive update on the progress of the Q3 Product Launch project, highlight some challenges we are facing, and outline the next steps for the upcoming weeks.

Firstly, I am pleased to report that the development team has successfully completed the beta version of the new application. Key features, including the user authentication module, dashboard analytics, and notification system, are functioning as expected. QA testing began last week and so far we have identified a few minor bugs related to the notification system and cross-browser compatibility. The team is actively working to resolve these issues, and I expect a new build to be available by the end of this week.

On the marketing front, our campaigns are progressing according to plan. Social media posts have been scheduled, and initial engagement metrics are promising. However, we did notice a slight delay in the production of promotional videos due to some scheduling conflicts with the external video team. We are working to expedite this and hope to have all materials ready for launch day.

From the sales perspective, our team has started outreach to key clients to provide previews of the new features. Initial feedback has been overwhelmingly positive, with several clients requesting additional demos and deeper technical specifications. We need to ensure that our support team is prepared to handle these requests efficiently and maintain clear communication channels with clients.

Regarding project risks, the main concern at this stage is integration with legacy systems. While our new APIs are well-documented, there is a dependency on client-side IT teams to update their infrastructure. We should consider scheduling follow-up meetings with them to mitigate potential delays.

Finally, I would like to schedule a team-wide meeting next Monday to discuss final preparations for the launch, review any remaining issues, and align on responsibilities. Please confirm your availability at your earliest convenience.

Thank you all for your continued hard work and dedication. Together, I am confident we can achieve a successful launch and exceed expectations.

Best regards,
John Smith
Product Manager
"""


```
