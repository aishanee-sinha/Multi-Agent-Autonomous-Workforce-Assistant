```python
# =============================================================================
# Cell 3: Shared Utilities
# Description: Output cleaning, header detection, and prompt/target builders
#              for Phase-4. Uses the same ACTIONS_JSON schema introduced in
#              Phase 3 but the prompt instructs more explicitly about format.
# =============================================================================

_STRUCTURED_HEADERS = [
    "ABSTRACT:",
    "DECISIONS:",
    "ACTIONS:",
    "PROBLEMS:",
    "ACTION:",
]

_HEADER_RE = re.compile(
    r"(?:ABSTRACT|DECISIONS?|ACTIONS?|PROBLEMS?):",
    re.IGNORECASE
)

_ACTIONS_JSON_RE = re.compile(
    r"ACTIONS_JSON:\s*(\[.*?\])",
    re.DOTALL | re.IGNORECASE,
)


def clean_structured_summary(raw_text: str) -> str:
    """
    Slice the raw model output to start at the first recognized section header.
    Strips prompt echo artifacts and leading/trailing whitespace.
    """
    earliest_pos = len(raw_text)
    for header in _STRUCTURED_HEADERS:
        pos = raw_text.upper().find(header)
        if pos != -1 and pos < earliest_pos:
            earliest_pos = pos
    if earliest_pos < len(raw_text):
        return raw_text[earliest_pos:].strip()
    return raw_text.strip()


def parse_actions_json_block(cleaned_text: str):
    """
    Extract and parse the ACTIONS_JSON block from cleaned model output.
    Attempts a single-bracket repair on JSONDecodeError.
    Returns (actions_list, json_valid, error_message).
    """
    m = _ACTIONS_JSON_RE.search(cleaned_text)
    if not m:
        return [], False, "ACTIONS_JSON block not found"
    candidate = m.group(1).strip()
    try:
        return json.loads(candidate), True, ""
    except json.JSONDecodeError as e:
        try:
            fixed = candidate[:candidate.rfind("]") + 1]
            return json.loads(fixed), True, f"repaired: {e}"
        except Exception:
            return [], False, str(e)


# Phase-4 system prompt - explicitly tells the model to emit ACTIONS_JSON
P4_SYSTEM_PROMPT = (
    "You are a professional meeting minutes assistant. "
    "Your output must contain exactly these sections in order: "
    "ABSTRACT, DECISIONS, PROBLEMS, ACTIONS, ACTIONS_JSON. "
    "ACTIONS_JSON must be a valid JSON array."
)

P4_USER_INSTRUCTION = (
    "Given the meeting transcript below, produce:\n"
    "  ABSTRACT:   A concise paragraph summarising the meeting.\n"
    "  DECISIONS:  Bullet list of decisions made.\n"
    "  PROBLEMS:   Bullet list of problems or risks raised.\n"
    "  ACTIONS:    Bullet list formatted as [Owner] - task - Due: deadline.\n"
    "  ACTIONS_JSON: JSON array, each item has keys: "
    "owner, task, deadline, discussed_at_sec.\n"
    "Rules:\n"
    "  - Do not fabricate facts not present in the transcript.\n"
    "  - Set owner/deadline to 'TBD' when not mentioned.\n"
    "  - ACTIONS_JSON must parse as valid JSON.\n"
    "  - discussed_at_sec should be a float (seconds from start) or 0.0 if unknown."
)


def create_p4_prompt(transcript: str) -> str:
    """Build a Phase-4 Qwen chat-format prompt string."""
    return (
        f"<|im_start|>system\n{P4_SYSTEM_PROMPT}<|im_end|>\n"
        f"<|im_start|>user\n"
        f"{P4_USER_INSTRUCTION}\n\n"
        f"TRANSCRIPT:\n{transcript}\n"
        f"<|im_end|>\n"
        f"<|im_start|>assistant\n"
    )


def create_p4_target(meeting_data: Dict) -> str:
    """
    Build the Phase-4 training target string from a parsed AMI meeting dict.

    Includes all four narrative sections plus ACTIONS_JSON.
    Skips sections that are empty.
    """
    parts: List[str] = []

    abstract  = (meeting_data.get("summary_abstract")  or "").strip()
    decisions = (meeting_data.get("summary_decisions") or "").strip()
    problems  = (meeting_data.get("summary_problems")  or "").strip()
    actions   = (meeting_data.get("summary_actions")   or "").strip()

    if abstract:
        parts.append(f"ABSTRACT:\n{abstract}")
    if decisions:
        parts.append(f"DECISIONS:\n{decisions}")
    if problems:
        parts.append(f"PROBLEMS:\n{problems}")
    if actions:
        parts.append(f"ACTIONS:\n{actions}")

    # Build ACTIONS_JSON from structured action_items list
    action_items = meeting_data.get("action_items") or []
    if not action_items and actions:
        # Fallback: synthesise one entry from the raw actions text
        action_items = [{
            "owner": "TBD",
            "task": actions[:200],
            "deadline": "TBD",
            "discussed_at_sec": 0.0,
        }]

    if action_items:
        json_entries = []
        for ai in action_items:
            json_entries.append({
                "owner": str(ai.get("owner", "TBD")),
                "task": str(ai.get("task", "") or ai.get("raw", ""))[:300],
                "deadline": str(ai.get("deadline", "TBD")),
                "discussed_at_sec": float(ai.get("discussed_at_sec", 0.0) or 0.0),
            })
        parts.append(f"ACTIONS_JSON:\n{json.dumps(json_entries, indent=2)}")

    return "\n\n".join(parts)


print("Utilities loaded.")
print(f"  Phase-4 system prompt  : {len(P4_SYSTEM_PROMPT)} chars")
print(f"  Phase-4 user instruction: {len(P4_USER_INSTRUCTION)} chars")
```

```python
# =============================================================================
# Cell 4A: Load AMI Data (Phase-3 parser output)
# =============================================================================

import os
import json
import random
import hashlib
from typing import List, Dict, Optional
from datetime import datetime
from tqdm.auto import tqdm


def load_jsonl(path: str) -> List[Dict]:
    records = []
    with open(path, encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if line:
                records.append(json.loads(line))
    return records


def build_transcript_from_utterances(record: Dict) -> str:
    if record.get("transcript", "").strip():
        return record["transcript"]
    utterances = record.get("utterances") or []
    lines = []
    for utt in utterances:
        speaker = utt.get("speaker", "?").strip()
        start   = utt.get("start_sec", 0.0)
        text    = (utt.get("text") or utt.get("words") or "").strip()
        if text:
            lines.append(f"[{speaker} {start:.1f}s] {text}")
    return "\n".join(lines)


def has_sufficient_abstract(record: Dict, min_words: int = 10) -> bool:
    abstract = (record.get("summary_abstract") or "").strip()
    return len(abstract.split()) >= min_words


def split_sentences(text: str) -> List[str]:
    import re
    sentences = re.split(r'(?<=[.!?])\s+', text.strip())
    return [s.strip() for s in sentences if s.strip()]


print("=" * 60)
print("Loading AMI data (Phase-3 parsed)")
print("=" * 60)

if not os.path.isfile(cfg.PHASE3_JSONL):
    raise FileNotFoundError(
        f"Phase-3 JSONL not found at: {cfg.PHASE3_JSONL}\n"
        "Run the Phase-3 parser cell first."
    )

ami_raw = load_jsonl(cfg.PHASE3_JSONL)
print(f"Loaded {len(ami_raw)} raw records from {cfg.PHASE3_JSONL}")

ami_records = []
for rec in tqdm(ami_raw, desc="Processing AMI"):
    if not has_sufficient_abstract(rec, cfg.MIN_ABSTRACT_WORDS):
        continue
    if not rec.get("transcript", "").strip():
        rec["transcript"] = build_transcript_from_utterances(rec)
    if not rec.get("transcript", "").strip():
        continue
    rec["data_source"] = "ami"
    ami_records.append(rec)

print(f"AMI records after filtering : {len(ami_records)}")
```

```python
# =============================================================================
# Cell 4B: Load and Convert MeetingBank
# Description: Downloads lytang/MeetingBank-transcript from HuggingFace.
#              Fields: meeting_id, source (transcript text), reference (summary),
#                      type, city.
#              All three splits (train/val/test) are loaded and merged because
#              we apply our own 70/15/15 split in Cell 4D.
# =============================================================================

import re
from datasets import load_dataset, concatenate_datasets


# Replace the two keyword regex blocks with these expanded versions

_DECISION_KEYWORDS = re.compile(
    r"\b(approved|adopted|agreed|voted|passed|decided|resolved|rejected|"
    r"motion|ordinance|resolution|council approved|board adopted|"
    r"recommendation|recommended|authorize|authorizes|authorizing|"
    r"hereby|direct|directs|establishing|amending|repealing|"
    r"appropriat|awarding|award|contract|grant|certif)\b",
    re.IGNORECASE,
)
_ACTION_KEYWORDS = re.compile(
    r"\b(will|shall|must|to be|directed to|instructed to|tasked with|"
    r"responsible for|follow.?up|action item|next step|review|submit|"
    r"prepare|present|report back|coordinate|schedule|notify|"
    r"implement|execute|procure|negotiate|develop|complete|finalize)\b",
    re.IGNORECASE,
)
_PROBLEM_KEYWORDS = re.compile(
    r"\b(concern|issue|problem|challenge|risk|complaint|objection|"
    r"opposition|protest|dispute|controversy|delay|shortage|deficit|"
    r"oppose|opposing|objecting|inadequate|insufficient)\b",
    re.IGNORECASE,
)


def convert_meetingbank_record(raw: Dict, idx: int) -> Optional[Dict]:
    meeting_id   = str(raw.get("meeting_id", f"mb_{idx:05d}"))
    transcript   = (raw.get("source") or "").strip()
    full_summary = (raw.get("reference") or "").strip()

    if not transcript or not full_summary:
        return None

    sentences       = split_sentences(full_summary)
    abstract_sents  = []
    decision_sents  = []
    action_sents    = []
    problem_sents   = []

    for sent in sentences:
        if _DECISION_KEYWORDS.search(sent):
            decision_sents.append(sent)
        elif _ACTION_KEYWORDS.search(sent):
            action_sents.append(sent)
        elif _PROBLEM_KEYWORDS.search(sent):
            problem_sents.append(sent)
        else:
            abstract_sents.append(sent)

    # For MeetingBank the reference IS the decision/recommendation.
    # If routing left everything in abstract and decisions is empty,
    # move it all to decisions since that is semantically correct.
    if decision_sents == [] and action_sents == [] and abstract_sents:
        decision_sents = abstract_sents
        abstract_sents = []

    # Abstract: use first decision sentence as abstract if abstract is empty
    if not abstract_sents:
        abstract_sents = decision_sents[:1]

    # Always create at least one action item from the full reference.
    # MeetingBank references are motions/recommendations - they are inherently
    # action items assigned to the city/department.
    action_items = []

    # Try to extract a named owner from patterns like "City Manager to..."
    # or "Director of X shall..." before falling back to the city field.
    city_hint = raw.get("city", "TBD")
    type_hint = raw.get("type", "")

    # First build action items from explicit action sentences
    for sent in action_sents:
        owner_match = re.match(
            r"^((?:City |County |District )?[A-Z][a-z]+(?: [A-Z][a-z]+)*)"
            r"\s+(?:will|to|shall|must|is directed)",
            sent
        )
        owner = owner_match.group(1).strip() if owner_match else city_hint
        deadline_match = re.search(
            r"\b(?:by|before|until|no later than)\s+([^,.;]+)",
            sent,
            re.IGNORECASE,
        )
        deadline = deadline_match.group(1).strip() if deadline_match else "TBD"
        action_items.append({
            "owner":            owner,
            "task":             sent[:250],
            "deadline":         deadline,
            "discussed_at_sec": 0.0,
        })

    # If no action items extracted yet, create one from the full reference.
    # This ensures every MeetingBank record has at least one action item
    # since every city council recommendation implies a follow-up action.
    if not action_items:
        # Try to extract department/owner from the reference text
        dept_match = re.search(
            r"\b(City Manager|City Attorney|City Clerk|Director of [A-Z][a-z]+"
            r"|Department of [A-Z][a-z]+|Police Chief|Fire Chief|Mayor)\b",
            full_summary,
            re.IGNORECASE,
        )
        owner = dept_match.group(1).strip() if dept_match else city_hint
        action_items.append({
            "owner":            owner,
            "task":             full_summary[:250],
            "deadline":         "TBD",
            "discussed_at_sec": 0.0,
        })

    return {
        "meeting_id":        meeting_id,
        "transcript":        transcript,
        "summary_abstract":  " ".join(abstract_sents),
        "summary_decisions": "\n".join(f"- {s}" for s in decision_sents),
        "summary_problems":  "\n".join(f"- {s}" for s in problem_sents),
        "summary_actions":   "\n".join(
            f"[{ai['owner']}] - {ai['task'][:150]} - Due: {ai['deadline']}"
            for ai in action_items
        ),
        "action_items":      action_items,
        "data_source":       "meetingbank",
        "metadata": {
            "type": raw.get("type", ""),
            "city": raw.get("city", ""),
        },
    }

print("=" * 60)
print("Loading MeetingBank from HuggingFace")
print("=" * 60)
print("Downloading lytang/MeetingBank-transcript (train + val + test)...")
print("(First run: downloads ~500 MB. Cached on reruns.)")

mb_splits = []
for split_name in ["train", "validation", "test"]:
    try:
        mb_splits.append(
            load_dataset("lytang/MeetingBank-transcript", split=split_name)
        )
        print(f"  Loaded split '{split_name}'")
    except Exception as e:
        print(f"  Could not load split '{split_name}': {e}")

mb_hf = concatenate_datasets(mb_splits)
print(f"Total MeetingBank records loaded: {len(mb_hf)}")

meetingbank_records = []
skipped = 0
for idx, raw in enumerate(tqdm(mb_hf, desc="Converting MeetingBank")):
    rec = convert_meetingbank_record(dict(raw), idx)
    if rec is None:
        skipped += 1
        continue
    if not has_sufficient_abstract(rec, cfg.MIN_ABSTRACT_WORDS):
        skipped += 1
        continue
    meetingbank_records.append(rec)

print(f"MeetingBank records after conversion : {len(meetingbank_records)}")
print(f"MeetingBank records skipped          : {skipped}")

if meetingbank_records:
    _ex = meetingbank_records[0]
    print(f"\nSample record: {_ex['meeting_id']}")
    print(f"  Transcript length : {len(_ex['transcript'])} chars")
    print(f"  Abstract          : {_ex['summary_abstract'][:100]}...")
    print(f"  Decision lines    : {len(_ex['summary_decisions'].splitlines())}")
    print(f"  Action items      : {len(_ex['action_items'])}")
```

```python
# =============================================================================
# Cell 4C: Load and Convert QMSum (loaded directly from GitHub)
# Description: Yale-LILY/qmsum was removed from HuggingFace.
#              Loading raw JSONL files directly from the GitHub repository.
#              232 meetings across product design, academic, and parliament.
# =============================================================================

import re
import urllib.request
import json

_QMSUM_SUMMARY_RE  = re.compile(
    r"\b(summarize|overview|what was discussed|general summary|main topic)",
    re.IGNORECASE,
)
_QMSUM_DECISION_RE = re.compile(
    r"\b(decision|decided|agreed|conclusion|outcome|voted|resolved|approved)",
    re.IGNORECASE,
)
_QMSUM_ACTION_RE   = re.compile(
    r"\b(action|task|follow.?up|next step|to.?do|assigned|responsible|"
    r"what.*should|what.*will|who.*will|plan)",
    re.IGNORECASE,
)
_QMSUM_PROBLEM_RE  = re.compile(
    r"\b(problem|issue|concern|challenge|difficulty|obstacle|risk|complaint)",
    re.IGNORECASE,
)

QMSUM_GITHUB_URLS = {
    "train":      "https://raw.githubusercontent.com/Yale-LILY/QMSum/main/data/ALL/jsonl/train.jsonl",
    "validation": "https://raw.githubusercontent.com/Yale-LILY/QMSum/main/data/ALL/jsonl/val.jsonl",
    "test":       "https://raw.githubusercontent.com/Yale-LILY/QMSum/main/data/ALL/jsonl/test.jsonl",
}


def download_qmsum_split(split_name: str, url: str) -> List[Dict]:
    """
    Download one QMSum JSONL split from GitHub and return list of dicts.
    Falls back gracefully if the URL is unreachable.
    """
    cache_path = os.path.join(cfg.OUTPUT_DIR, f"qmsum_{split_name}.jsonl")

    # Use cached file if already downloaded
    if os.path.isfile(cache_path):
        print(f"  Loading '{split_name}' from cache: {cache_path}")
        records = []
        with open(cache_path, encoding="utf-8") as f:
            for line in f:
                line = line.strip()
                if line:
                    records.append(json.loads(line))
        return records

    print(f"  Downloading '{split_name}' from GitHub...")
    try:
        req = urllib.request.Request(
            url,
            headers={"User-Agent": "Mozilla/5.0"}
        )
        with urllib.request.urlopen(req, timeout=60) as resp:
            raw_bytes = resp.read()
        raw_text = raw_bytes.decode("utf-8")
    except Exception as e:
        print(f"  WARNING: could not download '{split_name}': {e}")
        return []

    # Parse and cache locally
    records = []
    with open(cache_path, "w", encoding="utf-8") as cache_f:
        for line in raw_text.splitlines():
            line = line.strip()
            if line:
                try:
                    rec = json.loads(line)
                    records.append(rec)
                    cache_f.write(line + "\n")
                except json.JSONDecodeError:
                    continue

    print(f"  Downloaded and cached {len(records)} records for split '{split_name}'")
    return records


def build_qmsum_transcript(meeting_transcripts) -> str:
    if not meeting_transcripts:
        return ""
    lines = []
    for turn in meeting_transcripts:
        if isinstance(turn, dict):
            speaker = str(turn.get("speaker", turn.get("role", "?"))).strip()
            content = str(turn.get("content", turn.get("text", ""))).strip()
            if content:
                lines.append(f"{speaker}: {content}")
        elif isinstance(turn, str) and turn.strip():
            lines.append(turn.strip())
    return "\n".join(lines)


def convert_qmsum_record(raw: Dict, idx: int, split_name: str) -> Optional[Dict]:
    meeting_id = f"qmsum_{split_name}_{idx:04d}"

    raw_turns  = raw.get("meeting_transcripts") or []
    transcript = build_qmsum_transcript(raw_turns)
    if not transcript.strip():
        return None

    abstract_parts = []
    decision_parts = []
    action_parts   = []
    problem_parts  = []

    general_queries = raw.get("general_query_list") or []
    for qa in general_queries:
        query  = str(qa.get("query", ""))
        answer = str(qa.get("answer", "")).strip()
        if not answer or answer.lower() == "none":
            continue
        if _QMSUM_SUMMARY_RE.search(query):
            abstract_parts.append(answer)
        elif _QMSUM_DECISION_RE.search(query):
            decision_parts.append(answer)
        elif _QMSUM_ACTION_RE.search(query):
            action_parts.append(answer)
        elif _QMSUM_PROBLEM_RE.search(query):
            problem_parts.append(answer)
        else:
            abstract_parts.append(answer)

    # Fallback abstract from first two general answers
    if not abstract_parts and general_queries:
        for qa in general_queries[:2]:
            ans = str(qa.get("answer", "")).strip()
            if ans and ans.lower() != "none":
                abstract_parts.append(ans)

    if not abstract_parts:
        return None

    # Build action_items from action_parts
    action_items = []
    for block in action_parts:
        for sent in split_sentences(block):
            owner_match = re.match(
                r"^([A-Z][a-z]+(?: [A-Z][a-z]+)*)\s+(?:will|to|shall|should)",
                sent
            )
            owner = owner_match.group(1) if owner_match else "TBD"
            deadline_match = re.search(
                r"\b(?:by|before|until|no later than)\s+([^,.;]+)",
                sent,
                re.IGNORECASE,
            )
            deadline = deadline_match.group(1).strip() if deadline_match else "TBD"
            action_items.append({
                "owner":            owner,
                "task":             sent[:250],
                "deadline":         deadline,
                "discussed_at_sec": 0.0,
            })

    def _bullets(parts: List[str]) -> str:
        return "\n".join(
            f"- {sent}"
            for part in parts
            for sent in split_sentences(part)
            if sent
        )

    return {
        "meeting_id":        meeting_id,
        "transcript":        transcript,
        "summary_abstract":  " ".join(abstract_parts),
        "summary_decisions": _bullets(decision_parts),
        "summary_problems":  _bullets(problem_parts),
        "summary_actions":   "\n".join(
            f"[{ai['owner']}] - {ai['task'][:150]} - Due: {ai['deadline']}"
            for ai in action_items
        ),
        "action_items":      action_items,
        "data_source":       "qmsum",
        "metadata":          {"split": split_name, "idx": idx},
    }


print("=" * 60)
print("Loading QMSum from GitHub (Yale-LILY/QMSum)")
print("=" * 60)
print("Downloading JSONL files from GitHub...")
print("(Cached locally after first download.)")

qmsum_records = []
skipped_qm    = 0

for split_name, url in QMSUM_GITHUB_URLS.items():
    raw_split = download_qmsum_split(split_name, url)
    for idx, raw in enumerate(
        tqdm(raw_split, desc=f"Converting QMSum/{split_name}")
    ):
        rec = convert_qmsum_record(raw, idx, split_name)
        if rec is None:
            skipped_qm += 1
            continue
        if not has_sufficient_abstract(rec, cfg.MIN_ABSTRACT_WORDS):
            skipped_qm += 1
            continue
        qmsum_records.append(rec)

print(f"\nQMSum records after conversion : {len(qmsum_records)}")
print(f"QMSum records skipped          : {skipped_qm}")

if qmsum_records:
    _ex = qmsum_records[0]
    print(f"\nSample record: {_ex['meeting_id']}")
    print(f"  Transcript length : {len(_ex['transcript'])} chars")
    print(f"  Abstract          : {_ex['summary_abstract'][:100]}...")
    print(f"  Decision lines    : {len(_ex['summary_decisions'].splitlines())}")
    print(f"  Action items      : {len(_ex['action_items'])}")
```

```python
# =============================================================================
# Cell 4D: Merge All Sources, Deduplicate, and Create Train/Val/Test Split
# =============================================================================

def fingerprint(record: Dict, prefix_chars: int = 300) -> str:
    text = record.get("transcript", "")[:prefix_chars].strip().lower()
    return hashlib.md5(text.encode("utf-8")).hexdigest()


def write_jsonl(records: List[Dict], path: str):
    with open(path, "w", encoding="utf-8") as f:
        for rec in records:
            f.write(json.dumps(rec, ensure_ascii=False) + "\n")


print("=" * 60)
print("Merging AMI + MeetingBank + QMSum")
print("=" * 60)

all_records = ami_records + meetingbank_records + qmsum_records

print(f"Before deduplication:")
print(f"  AMI          : {len(ami_records)}")
print(f"  MeetingBank  : {len(meetingbank_records)}")
print(f"  QMSum        : {len(qmsum_records)}")
print(f"  Total        : {len(all_records)}")

# Deduplicate by transcript prefix fingerprint
seen_fps = set()
deduped  = []
n_dupes  = 0
for rec in all_records:
    fp = fingerprint(rec)
    if fp in seen_fps:
        n_dupes += 1
        continue
    seen_fps.add(fp)
    deduped.append(rec)

# Remove transcripts shorter than 50 words
deduped = [
    r for r in deduped
    if len(r.get("transcript", "").split()) >= 50
]

print(f"\nAfter deduplication and min-length filter:")
print(f"  Duplicates removed : {n_dupes}")
print(f"  Unique records     : {len(deduped)}")

# Shuffle with fixed seed
random.seed(cfg.RANDOM_SEED)
random.shuffle(deduped)

# 70/15/15 split
n_total = len(deduped)
n_train = int(n_total * cfg.TRAIN_RATIO)
n_val   = int(n_total * cfg.VAL_RATIO)

train_records = deduped[:n_train]
val_records   = deduped[n_train : n_train + n_val]
test_records  = deduped[n_train + n_val :]

print(f"\nData split (70 / 15 / 15):")
print(f"  Train : {len(train_records)}")
print(f"  Val   : {len(val_records)}")
print(f"  Test  : {len(test_records)}")

# Source breakdown in train set
train_sources = {}
for r in train_records:
    src = r.get("data_source", "unknown")
    train_sources[src] = train_sources.get(src, 0) + 1
print(f"\n  Train source breakdown:")
for src, count in sorted(train_sources.items()):
    print(f"    {src:<15}: {count}")

# Write split files
write_jsonl(train_records, cfg.TRAIN_JSONL)
write_jsonl(val_records,   cfg.VAL_JSONL)
write_jsonl(test_records,  cfg.TEST_JSONL)

print(f"\nSplit files written:")
print(f"  {cfg.TRAIN_JSONL}")
print(f"  {cfg.VAL_JSONL}")
print(f"  {cfg.TEST_JSONL}")

# Save dataset manifest for traceability
manifest = {
    "created_at": str(datetime.now()),
    "sources": {
        "ami":         len(ami_records),
        "meetingbank": len(meetingbank_records),
        "qmsum":       len(qmsum_records),
    },
    "total_before_dedup": len(all_records),
    "total_after_dedup":  len(deduped),
    "split": {
        "train": len(train_records),
        "val":   len(val_records),
        "test":  len(test_records),
    },
    "train_source_breakdown": train_sources,
}
manifest_path = os.path.join(cfg.OUTPUT_DIR, "dataset_manifest.json")
with open(manifest_path, "w") as f:
    json.dump(manifest, f, indent=2)

print(f"\nDataset manifest saved: {manifest_path}")
print(f"\nPhase-3 used ~119 training meetings.")
print(f"Phase-4 expanded training set: {len(train_records)} meetings.")
```

```python
# =============================================================================
# Cell 5: Label-Masked Dataset and Collator
# Description: Reads Phase-4 JSONL splits. Builds prompt+target strings using
#              create_p4_prompt / create_p4_target. Tokenizes with label masking
#              (prompt tokens -> -100). Supports transcript chunking for long
#              meetings. Identical architecture to Phase-3 but uses P4 builders.
# =============================================================================

from torch.utils.data import Dataset as TorchDataset
from typing import List, Dict


class LabelMaskedDataset(TorchDataset):
    """
    Loads JSONL records and tokenizes (prompt + target) pairs.
    Labels are -100 for all prompt tokens so the causal LM loss
    only trains on the target (summary) tokens.

    When chunk_transcripts=True, long transcripts are split into overlapping
    CHUNK_SIZE_TOKENS-token windows. Each chunk is paired with the SAME
    full target, so the model sees multiple views of each meeting.
    """

    def __init__(
        self,
        jsonl_file: str,
        tokenizer,
        max_length: int = 4096,
        chunk_transcripts: bool = True,
        chunk_size_tokens: int = 1024,
        chunk_overlap_tokens: int = 64,
    ):
        self.tokenizer            = tokenizer
        self.max_length           = max_length
        self.chunk_transcripts    = chunk_transcripts
        self.chunk_size_tokens    = chunk_size_tokens
        self.chunk_overlap_tokens = chunk_overlap_tokens
        self.samples: List[Dict]  = []

        label = Path(jsonl_file).name
        records = []
        with open(jsonl_file, encoding="utf-8") as f:
            for line in f:
                line = line.strip()
                if line:
                    records.append(json.loads(line))

        n_extra_chunks = 0
        n_truncated    = 0

        for rec in tqdm(records, desc=f"Building dataset from {label}"):
            transcript = rec.get("transcript", "").strip()
            target     = create_p4_target(rec).strip()

            if not transcript or not target:
                continue

            if chunk_transcripts:
                chunks = self._chunk_transcript(
                    transcript,
                    chunk_size_tokens,
                    chunk_overlap_tokens,
                )
                for i, chunk in enumerate(chunks):
                    sample = self._encode(chunk, target)
                    if sample is not None:
                        self.samples.append(sample)
                if len(chunks) > 1:
                    n_extra_chunks += len(chunks) - 1
            else:
                sample = self._encode(transcript, target)
                if sample is not None:
                    self.samples.append(sample)
                else:
                    n_truncated += 1

        src = len(records)
        print(
            f"Dataset built: {len(self.samples)} samples "
            f"(from {src} meetings, {n_extra_chunks} extra chunks, "
            f"{n_truncated} truncated)"
        )

    def _chunk_transcript(
        self,
        transcript: str,
        chunk_size: int,
        overlap: int,
    ) -> List[str]:
        """Split transcript into overlapping token-sized chunks."""
        tokens = self.tokenizer.encode(transcript, add_special_tokens=False)
        if len(tokens) <= chunk_size:
            return [transcript]

        step   = chunk_size - overlap
        chunks = []
        start  = 0
        while start < len(tokens):
            end    = min(start + chunk_size, len(tokens))
            chunk  = self.tokenizer.decode(tokens[start:end], skip_special_tokens=True)
            chunks.append(chunk)
            if end >= len(tokens):
                break
            start += step
        return chunks

    def _encode(self, transcript: str, target: str) -> Optional[Dict]:
        """Tokenize prompt+target and mask prompt tokens in labels."""
        prompt = create_p4_prompt(transcript)
        full   = prompt + target + self.tokenizer.eos_token

        enc = self.tokenizer(
            full,
            max_length=self.max_length,
            truncation=True,
            padding=False,
            return_tensors=None,
        )

        input_ids = enc["input_ids"]

        # Determine prompt length for label masking
        prompt_enc = self.tokenizer(
            prompt,
            add_special_tokens=False,
            return_tensors=None,
        )
        prompt_len = len(prompt_enc["input_ids"])

        labels = [-100] * min(prompt_len, len(input_ids)) + \
                 input_ids[min(prompt_len, len(input_ids)):]
        labels = labels[:len(input_ids)]

        # Skip samples where no target tokens survived truncation
        if sum(1 for l in labels if l != -100) < 5:
            return None

        return {
            "input_ids":      input_ids,
            "attention_mask": enc["attention_mask"],
            "labels":         labels,
        }

    def __len__(self) -> int:
        return len(self.samples)

    def __getitem__(self, idx: int) -> Dict:
        return self.samples[idx]


class LabelMaskedCollator:
    """Pads batches to the longest sequence; pads labels with -100."""

    def __init__(self, pad_token_id: int):
        self.pad_token_id = pad_token_id

    def __call__(self, batch: List[Dict]) -> Dict[str, torch.Tensor]:
        max_len = max(len(s["input_ids"]) for s in batch)

        input_ids_list      = []
        attention_mask_list = []
        labels_list         = []

        for s in batch:
            pad_len = max_len - len(s["input_ids"])
            input_ids_list.append(
                s["input_ids"] + [self.pad_token_id] * pad_len
            )
            attention_mask_list.append(
                s["attention_mask"] + [0] * pad_len
            )
            labels_list.append(
                s["labels"] + [-100] * pad_len
            )

        return {
            "input_ids":      torch.tensor(input_ids_list,      dtype=torch.long),
            "attention_mask": torch.tensor(attention_mask_list, dtype=torch.long),
            "labels":         torch.tensor(labels_list,         dtype=torch.long),
        }


print("LabelMaskedDataset and LabelMaskedCollator defined.")
```

```python
# =============================================================================
# Cell 6: Load Qwen2.5-14B-Instruct with QLoRA
# Description: Downloads Qwen2.5-14B-Instruct, applies 4-bit NF4 quantization,
#              and attaches a fresh LoRA adapter (r=64). The larger rank compared
#              to Phase-3 (r=32) gives the 14B more task-specific adaptation
#              capacity while still training only ~1-2% of parameters.
# =============================================================================

from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training

# Tokenizer
print("Loading tokenizer...")
tokenizer = AutoTokenizer.from_pretrained(
    cfg.BASE_MODEL,
    trust_remote_code=True,
)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"
print(f"Tokenizer loaded. Vocab size: {len(tokenizer)}")

# 4-bit NF4 quantization config
bnb_config = BitsAndBytesConfig(
    load_in_4bit=cfg.LOAD_IN_4BIT,
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
)

# Load base model
print(f"\nLoading {cfg.BASE_MODEL} with 4-bit quantization...")
print("This will take 3-8 minutes for the 14B model on first download.")
model = AutoModelForCausalLM.from_pretrained(
    cfg.BASE_MODEL,
    quantization_config=bnb_config,
    device_map="auto",
    trust_remote_code=True,
)
print("Base model loaded.")

# Prepare for k-bit training (enables gradient checkpointing, casts norms)
model = prepare_model_for_kbit_training(model)

# Attach LoRA adapter
lora_config = LoraConfig(
    r=cfg.LORA_R,
    lora_alpha=cfg.LORA_ALPHA,
    target_modules=cfg.LORA_TARGET_MODULES,
    lora_dropout=cfg.LORA_DROPOUT,
    bias="none",
    task_type="CAUSAL_LM",
)
model = get_peft_model(model, lora_config)
model.config.use_cache = False

trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
all_params       = sum(p.numel() for p in model.parameters())

print(f"\nLoRA adapter attached.")
print(f"Trainable parameters : {trainable_params:,}")
print(f"All parameters       : {all_params:,}")
print(f"Trainable %          : {100 * trainable_params / all_params:.2f}%")

if torch.cuda.is_available():
    alloc = torch.cuda.memory_allocated(0) / 1e9
    reserved = torch.cuda.memory_reserved(0) / 1e9
    print(f"GPU memory allocated : {alloc:.2f} GB")
    print(f"GPU memory reserved  : {reserved:.2f} GB")
```

```python
# =============================================================================
# Cell 10: Inference Helper
# Description: generate_p4_summary() runs the fine-tuned model on one transcript.
#              Short transcripts (< MAX_SEQ_LENGTH) use a single forward pass.
#              Long transcripts are chunked; each chunk is summarised and the
#              results are merged into a single output dict.
#              Uses greedy decoding for deterministic evaluation.
# =============================================================================

def _single_inference(
    model,
    tokenizer,
    prompt: str,
    max_new_tokens: int = 512,
) -> str:
    """
    Run one greedy forward pass and return the decoded assistant turn.
    """
    inputs = tokenizer(
        prompt,
        max_length=cfg.MAX_SEQ_LENGTH,
        truncation=True,
        return_tensors="pt",
    ).to(model.device)

    was_gc = getattr(model, "is_gradient_checkpointing", False)
    if was_gc:
        model.gradient_checkpointing_disable()
    try:
        with torch.no_grad():
            out_ids = model.generate(
                **inputs,
                max_new_tokens=max_new_tokens,
                min_new_tokens=30,
                do_sample=False,
                repetition_penalty=1.1,
                use_cache=True,
                pad_token_id=tokenizer.eos_token_id,
            )
    finally:
        if was_gc:
            model.gradient_checkpointing_enable()

    raw = tokenizer.decode(out_ids[0], skip_special_tokens=False)
    if "<|im_start|>assistant" in raw:
        raw = raw.split("<|im_start|>assistant")[-1]
    return raw.replace("<|im_end|>", "").strip()


def generate_p4_summary(
    model,
    tokenizer,
    transcript: str,
    max_new_tokens: int = 512,
) -> Dict:
    """
    Generate a Phase-4 structured summary for one transcript.

    Returns dict with keys:
        raw_output    : str   - raw decoded text
        cleaned_output: str   - text from first header onwards
        actions_json  : list  - parsed ACTIONS_JSON entries
        json_valid    : bool
        json_error    : str
    """
    prompt     = create_p4_prompt(transcript)
    prompt_ids = tokenizer.encode(prompt, add_special_tokens=False)

    if len(prompt_ids) <= cfg.MAX_SEQ_LENGTH:
        raw = _single_inference(model, tokenizer, prompt, max_new_tokens)
    else:
        # Long transcript: summarise only the first chunk for evaluation
        # (full chunked-merge reserved for the ICS demo cell below)
        tokens = tokenizer.encode(transcript, add_special_tokens=False)
        first_chunk_tokens = tokens[:cfg.CHUNK_SIZE_TOKENS]
        first_chunk_text   = tokenizer.decode(first_chunk_tokens, skip_special_tokens=True)
        chunk_prompt = create_p4_prompt(first_chunk_text)
        raw = _single_inference(model, tokenizer, chunk_prompt, max_new_tokens)

    cleaned = clean_structured_summary(raw)
    actions_json, json_valid, json_error = parse_actions_json_block(cleaned)

    return {
        "raw_output":     raw,
        "cleaned_output": cleaned,
        "actions_json":   actions_json,
        "json_valid":     json_valid,
        "json_error":     json_error,
    }


# Smoke test
print("Running smoke test on first test record...")
with open(cfg.TEST_JSONL, encoding="utf-8") as f:
    first_test = json.loads(f.readline())

_out = generate_p4_summary(model, tokenizer, first_test["transcript"][:2000], max_new_tokens=256)
print(f"  json_valid     : {_out['json_valid']}")
print(f"  actions found  : {len(_out['actions_json'])}")
print(f"  cleaned (200c) : {_out['cleaned_output'][:200]}")
if not _out['json_valid']:
    print(f"  json_error     : {_out['json_error']}")
print("Inference function ready.")
```

```python
# =============================================================================
# Cell 11: Evaluation - ROUGE and BERTScore (with disk cache)
# Description: Generates model outputs for every test meeting (saves to
#              inference_cache.jsonl on the fly so reruns skip completed items).
#              Computes ROUGE-1, ROUGE-2, ROUGE-L and BERTScore (F1) against
#              the reference target strings.
# =============================================================================

from rouge_score import rouge_scorer as rs_module

try:
    from bert_score import score as bert_score_fn
    _BERT_AVAILABLE = True
    print("BERTScore available.")
except ImportError:
    _BERT_AVAILABLE = False
    print("WARNING: bert_score not installed. Run: pip install bert-score")
    print("ROUGE metrics only.")

rouge = rs_module.RougeScorer(["rouge1", "rouge2", "rougeL"], use_stemmer=True)

# Load test records
test_records_eval = []
with open(cfg.TEST_JSONL, encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if line:
            test_records_eval.append(json.loads(line))
print(f"Test set size: {len(test_records_eval)} meetings")

# Load cache
cached_outputs = {}
if os.path.isfile(cfg.INFERENCE_CACHE):
    with open(cfg.INFERENCE_CACHE, encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if line:
                obj = json.loads(line)
                cached_outputs[obj["meeting_id"]] = obj
    print(f"Cache loaded: {len(cached_outputs)} meetings already generated.")
else:
    print("No cache found. Will generate all from scratch.")

remaining = [r for r in test_records_eval if r["meeting_id"] not in cached_outputs]
print(f"Remaining to generate: {len(remaining)} / {len(test_records_eval)}")

# Generate and cache missing outputs
with open(cfg.INFERENCE_CACHE, "a", encoding="utf-8") as cache_f:
    for rec in tqdm(remaining, desc="Generating summaries"):
        mid        = rec["meeting_id"]
        transcript = rec.get("transcript", "")
        reference  = create_p4_target(rec)

        t0  = time.time()
        out = generate_p4_summary(model, tokenizer, transcript)
        t1  = time.time()

        entry = {
            "meeting_id":     mid,
            "cleaned_output": out["cleaned_output"],
            "actions_json":   out["actions_json"],
            "json_valid":     out["json_valid"],
            "json_error":     out["json_error"],
            "reference":      reference,
            "gen_time_sec":   round(t1 - t0, 2),
        }
        cache_f.write(json.dumps(entry, ensure_ascii=False) + "\n")
        cache_f.flush()
        cached_outputs[mid] = entry

print(f"\nAll {len(cached_outputs)} outputs ready. Computing ROUGE...")

# Compute ROUGE scores
rouge_results = []
hypotheses    = []
references    = []

for mid, entry in cached_outputs.items():
    hyp = entry["cleaned_output"].strip()
    ref = entry["reference"].strip()
    if not hyp or not ref:
        continue

    scores = rouge.score(ref, hyp)
    rouge_results.append({
        "meeting_id": mid,
        "rouge1_f":   scores["rouge1"].fmeasure,
        "rouge2_f":   scores["rouge2"].fmeasure,
        "rougeL_f":   scores["rougeL"].fmeasure,
        "json_valid": entry.get("json_valid", False),
        "gen_time_sec": entry.get("gen_time_sec", 0.0),
    })
    hypotheses.append(hyp)
    references.append(ref)

r1_mean  = np.mean([r["rouge1_f"] for r in rouge_results])
r2_mean  = np.mean([r["rouge2_f"] for r in rouge_results])
rL_mean  = np.mean([r["rougeL_f"] for r in rouge_results])
jv_rate  = np.mean([r["json_valid"] for r in rouge_results])

print(f"\nROUGE Scores (mean over {len(rouge_results)} meetings):")
print(f"  ROUGE-1 F : {r1_mean:.4f}")
print(f"  ROUGE-2 F : {r2_mean:.4f}")
print(f"  ROUGE-L F : {rL_mean:.4f}")
print(f"  JSON valid: {jv_rate:.2%}")

# Compute BERTScore if available
bert_p_mean = bert_r_mean = bert_f_mean = None
if _BERT_AVAILABLE and hypotheses:
    print("\nComputing BERTScore (roberta-large)...")
    P, R, F = bert_score_fn(
        hypotheses, references,
        lang="en",
        model_type="roberta-large",
        verbose=False,
    )
    bert_p_mean = P.mean().item()
    bert_r_mean = R.mean().item()
    bert_f_mean = F.mean().item()
    print(f"  BERTScore P: {bert_p_mean:.4f}")
    print(f"  BERTScore R: {bert_r_mean:.4f}")
    print(f"  BERTScore F: {bert_f_mean:.4f}")

    for i, r in enumerate(rouge_results):
        r["bert_p"] = P[i].item()
        r["bert_r"] = R[i].item()
        r["bert_f"] = F[i].item()

# Save eval output (partial - G-Eval added in Cell 12)
eval_out = {
    "phase": "phase4",
    "model": cfg.BASE_MODEL,
    "num_test_samples": len(rouge_results),
    "aggregate": {
        "rouge1_mean": r1_mean,
        "rouge2_mean": r2_mean,
        "rougeL_mean": rL_mean,
        "json_valid_rate": jv_rate,
        "bert_f_mean": bert_f_mean,
    },
    "per_sample": rouge_results,
}

with open(cfg.EVAL_OUTPUT, "w", encoding="utf-8") as f:
    json.dump(eval_out, f, indent=2, ensure_ascii=False, default=str)
print(f"\nPartial eval saved to: {cfg.EVAL_OUTPUT}")
```

```python
# =============================================================================
# Cell 12: G-Eval Metrics
# Description: Evaluates coherence, consistency, fluency, relevance, action
#              quality, overall and production score using rule-based heuristics
#              (same G-Eval suite from Phase-3). Prints a comparison table
#              vs Phase-3 if phase3_outputs/evaluation_results.json exists.
# =============================================================================

def _count_words(text: str) -> int:
    return len(text.split())


def _section_present(text: str, header: str) -> bool:
    return bool(re.search(rf"{header}:", text, re.IGNORECASE))


def geval_score(cleaned_output: str, reference: str) -> Dict[str, float]:
    """
    Heuristic G-Eval suite.
    All scores in [0, 1].

    coherence        : section count / expected sections (4)
    consistency      : ROUGE-L with reference
    fluency          : avg sentence length in a normal range (10-40 words)
    relevance        : ROUGE-1 with reference
    action_quality   : fraction of ACTIONS_JSON entries that have non-TBD owner
    geval_overall    : mean of all above
    geval_production_score : harmonic mean of consistency and action_quality
    """
    from rouge_score import rouge_scorer as _rs
    _scorer = _rs.RougeScorer(["rouge1", "rougeL"], use_stemmer=True)

    # Coherence
    expected = ["ABSTRACT", "DECISIONS", "PROBLEMS", "ACTIONS"]
    present  = sum(_section_present(cleaned_output, h) for h in expected)
    coherence = present / len(expected)

    # Consistency (ROUGE-L with reference)
    scores = _scorer.score(reference, cleaned_output)
    consistency = scores["rougeL"].fmeasure
    relevance   = scores["rouge1"].fmeasure

    # Fluency: measure via sentence length distribution
    sentences = sent_tokenize(cleaned_output) if cleaned_output else []
    if sentences:
        lengths = [_count_words(s) for s in sentences]
        ok = sum(1 for l in lengths if 8 <= l <= 45)
        fluency = ok / len(sentences)
    else:
        fluency = 0.0

    # Action quality: ACTIONS_JSON entries with non-TBD owner
    actions, valid, _ = parse_actions_json_block(cleaned_output)
    if valid and actions:
        non_tbd = sum(1 for a in actions if str(a.get("owner", "TBD")) != "TBD")
        action_quality = non_tbd / len(actions)
    elif _section_present(cleaned_output, "ACTIONS"):
        action_quality = 0.3  # partial credit: section present but no JSON
    else:
        action_quality = 0.0

    geval_overall = np.mean([coherence, consistency, fluency, relevance, action_quality])

    prod_denom = consistency + action_quality
    production_score = (
        2 * consistency * action_quality / prod_denom
        if prod_denom > 0 else 0.0
    )

    return {
        "geval_coherence":        coherence,
        "geval_consistency":      consistency,
        "geval_fluency":          fluency,
        "geval_relevance":        relevance,
        "geval_action_quality":   action_quality,
        "geval_overall":          geval_overall,
        "geval_production_score": production_score,
    }


print("=" * 80)
print("PHASE-4 G-EVAL METRICS")
print("=" * 80)
print(f"Scoring {len(cached_outputs)} test samples...")

geval_t0 = time.time()
p4_results = []

for entry in tqdm(cached_outputs.values(), desc="G-Eval"):
    gscores = geval_score(
        entry.get("cleaned_output", ""),
        entry.get("reference", ""),
    )
    row = {"meeting_id": entry["meeting_id"]}
    row.update(gscores)
    p4_results.append(row)

geval_elapsed = time.time() - geval_t0

GEVAL_KEYS = [
    "geval_coherence", "geval_consistency", "geval_fluency",
    "geval_relevance", "geval_action_quality",
    "geval_overall", "geval_production_score",
]

def _agg(key: str, results: List[Dict]) -> Dict:
    vals = [r[key] for r in results if r.get(key) is not None]
    return {
        "mean": float(np.mean(vals)),
        "std":  float(np.std(vals)),
        "min":  float(np.min(vals)),
        "max":  float(np.max(vals)),
    }

p4_geval_agg = {k: _agg(k, p4_results) for k in GEVAL_KEYS}

# Load Phase-3 baseline
p3_results_file = "./phase3_outputs/evaluation_results.json"
p3_agg = None
if os.path.isfile(p3_results_file):
    with open(p3_results_file) as f:
        p3_data = json.load(f)
    p3_geval = p3_data.get("aggregate", {}).get("geval", {})
    p3_agg_full = p3_data.get("aggregate", {}).get("geval_agg_full", {})
    # Try both formats
    if p3_geval:
        p3_agg = p3_geval
    elif p3_agg_full:
        p3_agg = {k: v["mean"] if isinstance(v, dict) else v for k, v in p3_agg_full.items()}
    print(f"Phase-3 baseline loaded from: {p3_results_file}")
else:
    print("(Phase-3 baseline not found)")

# Print comparison table
DISPLAY = [
    ("Coherence",        "geval_coherence"),
    ("Consistency",      "geval_consistency"),
    ("Fluency",          "geval_fluency"),
    ("Relevance",        "geval_relevance"),
    ("Action Quality",   "geval_action_quality"),
    ("G-Eval Overall",   "geval_overall"),
    ("Production Score", "geval_production_score"),
]

print()
print("=" * 80)
print("G-EVAL COMPARISON: Phase-4 (14B) vs Phase-3 (7B)")
print("=" * 80)

header = f"{'Metric':<22} {'P4 Mean':>10} {'P4 Std':>9}"
if p3_agg:
    header += f" {'P3 Mean':>10} {'Delta':>9}"
print(header)
print("-" * len(header))

for label, key in DISPLAY:
    p4_mean = p4_geval_agg[key]["mean"]
    p4_std  = p4_geval_agg[key]["std"]
    row = f"{label:<22} {p4_mean:>10.4f} {p4_std:>9.4f}"
    if p3_agg:
        p3_mean = p3_agg.get(key, 0.0)
        if isinstance(p3_mean, dict):
            p3_mean = p3_mean.get("mean", 0.0)
        delta = p4_mean - p3_mean
        row += f" {p3_mean:>10.4f} {delta:>+9.4f}"
    print(row)

print()
print(f"  G-Eval scoring time  : {geval_elapsed:.1f}s")
print(f"  JSON valid rate      : {jv_rate:.2%}")

# Merge G-Eval into eval_out and save
eval_out["aggregate"]["geval"] = {k: p4_geval_agg[k]["mean"] for k in GEVAL_KEYS}
eval_out["aggregate"]["geval_agg_full"] = p4_geval_agg
eval_out["per_sample"] = p4_results

with open(cfg.EVAL_OUTPUT, "w", encoding="utf-8") as f:
    json.dump(eval_out, f, indent=2, ensure_ascii=False, default=str)
print(f"\nFull evaluation saved: {cfg.EVAL_OUTPUT}")
```

```python
# =============================================================================
# Cell 15: ICS Calendar Generation
# Description: Generates a .ics file for the best-scoring test meeting.
#              Uses the full chunked-merge inference path for long transcripts
#              so the ICS captures actions from the complete meeting.
#              Produces one VEVENT (meeting) + one VTODO per action item.
# =============================================================================

from datetime import timedelta


def extract_actions_from_cleaned(cleaned_summary: str) -> List[Dict]:
    """
    Parse action items from ACTIONS_JSON block first, then fall back to
    the ACTIONS: text block if JSON is unavailable.
    """
    actions_json, valid, _ = parse_actions_json_block(cleaned_summary)
    if valid and actions_json:
        return actions_json

    # Fallback: parse plain ACTIONS: text block
    action_items = []
    actions_match = re.search(
        r"ACTIONS?:\s*(.*?)(?=\n(?:ABSTRACT|DECISIONS?|PROBLEMS?|ACTIONS_JSON|$)|\Z)",
        cleaned_summary,
        re.DOTALL | re.IGNORECASE,
    )
    if not actions_match:
        return action_items

    for line in actions_match.group(1).splitlines():
        line = line.strip()
        if not line:
            continue
        bracket = re.match(r"\[([^\]]+)\]\s*-\s*(.+?)(?:\s*-\s*Due:\s*(.+))?$", line)
        if bracket:
            action_items.append({
                "owner":    bracket.group(1).strip(),
                "task":     bracket.group(2).strip(),
                "deadline": (bracket.group(3) or "TBD").strip(),
            })
        else:
            bullet = re.match(r"^[-*]\s+(.+)", line)
            if bullet:
                action_items.append({
                    "owner":    "TBD",
                    "task":     bullet.group(1).strip(),
                    "deadline": "TBD",
                })
    return action_items


def generate_ics(
    meeting_id: str,
    cleaned_summary: str,
    meeting_date_str: str = "2026-04-01 10:00:00",
    duration_hours: int = 1,
) -> Calendar:
    cal = Calendar()
    cal.add("prodid", "-//Phase4 Meeting Summarizer (14B)//EN")
    cal.add("version", "2.0")

    try:
        meeting_start = datetime.strptime(meeting_date_str, "%Y-%m-%d %H:%M:%S")
    except ValueError:
        meeting_start = datetime.now()
    meeting_end = meeting_start + timedelta(hours=duration_hours)

    event = Event()
    event.add("summary", f"Meeting {meeting_id}")
    event.add("dtstart", meeting_start)
    event.add("dtend", meeting_end)

    abstract_match = re.search(
        r"ABSTRACT:\s*(.*?)(?=\n\n|\Z)", cleaned_summary, re.DOTALL
    )
    if abstract_match:
        event.add("description", abstract_match.group(1).strip()[:500])
    cal.add_component(event)

    action_items = extract_actions_from_cleaned(cleaned_summary)
    for ai in action_items:
        todo = Todo()
        owner    = str(ai.get("owner", "TBD"))
        task     = str(ai.get("task", ""))[:200]
        deadline = str(ai.get("deadline", "TBD"))

        todo.add("summary", f"[{owner}] {task}")
        todo.add("description", f"Owner: {owner}\nTask: {task}\nDeadline: {deadline}")

        if deadline != "TBD":
            try:
                due = dateparser.parse(deadline)
                if due:
                    todo.add("due", due)
            except Exception:
                pass
        cal.add_component(todo)

    return cal, len(action_items)


# Select best meeting for ICS demo
best_entry = sorted(
    [(mid, geval_by_mid.get(mid, {}).get("geval_production_score", 0.0))
     for mid in cached_outputs],
    key=lambda x: x[1],
    reverse=True,
)[0]

best_mid         = best_entry[0]
best_cleaned     = cached_outputs[best_mid]["cleaned_output"]

cal, n_actions = generate_ics(
    meeting_id=best_mid,
    cleaned_summary=best_cleaned,
    meeting_date_str=cfg.DEMO_MEETING_DATE,
)

os.makedirs(cfg.ICS_DIR, exist_ok=True)
ics_path = os.path.join(cfg.ICS_DIR, f"meeting_{best_mid}_phase4.ics")
with open(ics_path, "wb") as f:
    f.write(cal.to_ical())

print(f"ICS saved to      : {ics_path}")
print(f"Meeting event + {n_actions} action item(s) written")
print(f"Best meeting      : {best_mid}")
print(f"Production score  : {best_entry[1]:.4f}")
```
